# 트래픽
- 주소 체계 정리

## 네트워크 주소
- MAC Address(2계층 주소)
- IP Address(3계층 주소)
- Port Number(4계층 주소)
- FQDN(7계층 주소)

### 논리석 주소(3계층 주소)
- IP Address 구성: Network ID + Host ID
- Subnet Mask 기능: IP Address의 Network ID와 Host ID 구분

### 내부망과 외부망
- 송신자: 데이터를 보내는 측/ 수신자: 데이터를 받는 측
- 오늘은 A를 수신자로 설정

- 내부망: 송수신지의 네트워크 ID가 **동일**(송신자가 수신자는 내부망에 있다는 것을 결정! 결정하는 사람은 송신자!!)
- 외부망: 송수신지의 네트워크 ID가 **다름**

### 물리적 주소(2계층)
- 네트워크 인터페이스 카드(NIC) 또는 이더넷 카드
- 데이터링크 계층의 MAC 계층에 의해 사용되는 48비트의 하드웨어 주소
- 여러 종류의 LAN카드
    - Token ring
    - Ethernet
    - FDDI 등등
    - 각각 LAN카드 종류마다 주소가 다르게 부여된다.
- 맥주소는 **16진수의 12자리**로 구성되어있다.
    - 앞 6자리는 제조회사 식별번호, 뒤 6자리는 카드의 일련번호

- IP 충돌: 두 컴퓨터의 주소가 동일한 것
    - 수동으로 해결할 수 있음
- MAC 주소 충돌: 두 컴퓨터의 맥 주소가 동일한 것. 
    - 수동으로 설정하는 것이 아니라 발생하기 쉽지 않다.
    - LAN 카드의 제조회사 식별번호를 복제하여 중복하는 경우가 발생했었음
    - MAC Address 충돌이 더 심각한 문제!
    - **네트워크 보안에서는 MAC Address가 중요하다**
    


### FQDN(Fully Qualified Domain Name, 7계층 주소)
- Host Name + Domain Name
- 문자 주소

- 클라이언트가 존재해야 서버가 존재한다!
    - 클라이언트의 호출이 있어야 서버가 존재!!

- Host Name: www, ftp, smtp 등
- Domain Name: test.com...



### MAC/IP/FQDN Address
- 서버는 여러 개의 주소를 가질 수 있다.(3개로 시작)
    - IP, MAC, FQDN 등으로 호출이 가능하다.
    - 문자형 주소는 서비스적인 주소이다. 사람이 기억하기 쉬운 주소로 클라이언트 친화적 주소.
    - 통신할 때 필요한 주소는 IP, MAC 주소이다.

### DNS & ARP
- 네트워크 주소 서비스 관련 프로토콜
- 문자 주소의 IP 주소를 알려주는 DNS, IP 주소의 MAC 주소를 알려주는 것이 ARP

- DNS: FQDN(문자 주소) <--> IP   #조회(lookup이라고 함)
- ARP: IP --> MAC   #resolution
- 한 번에 문자 주소를 맥주소로 바꿔주는 프로토콜은 존재하지 않는다! --> 단계별로 접근
- RARP: MAC --> IP

- 네트워크 공격에서는 가장 기본적인 서비스인 DNS와 ARP를 통한 공격이 많다!
    - DNS는 애플리케이션 계층으로 ARP에 대해서 배움(네트워크 계층)
    




### Port Number(포트 번호, 4계층)
- 데이터 송수신 번호
- 서비스 번호 또는 애플리케이션 번호
애플리케이션에 부착해 전송
- 포트번호 0번은 사용하지 않음
- 포트 번호는 프로그램 번호!
    - 클라이언트용 프로그램: 크롬(웹 클라이언트)
    - 서버용 프로그램: Apache(웹 서버)

- well known port: 1~1023, 웹 서버용 번호. ICANN가 특정 애플리케이션용으로 미리 예약한 포트 번호. 반영구적, 정적 성향이 강함

- registered port: 1024~49151: 신규 애플리케이션에 등록해서 사용할 수 있는 포트. 기업이나 개발자가 등록한 포트, 특정 앱용으로 사용.
    - 슈퍼유저 권한 없이 사용 가능: 일반 사용자도 접근 가능.
    - 공식 등록 필요: IANA에 신청해서 등록
    - 서비스 식별에 사용: 예를 들어, MySQL은 3306번 포트를 사용하고, Microsoft SQL Server는 1433번 포트를 사용

- dynamic port: 클라이언트에 부여하는 번호로 운영체제가 부여하는 동적포트 번호, 개인적인 목적으로 사용할 수 있는 포트이다. 크롬을 열었을 때, 50010, 50020 등으로 할당된다.
    - 작업을 종료할 시 포트 번호는 반환되어 재할당이 가능해진다.(동적 성향)

- 서버 설정
    1. 아파치와 같은 서버 프로그램 설치
    2. 환경 설정에서 포트 번호 설정(서버 담당자가 변경 가능)
        - well known 포트는 클라이언트들에게 잘 알려진 포트 번호로, 담당자가 포트 번호를 바꾸면 클라이언트는 알려진 포트 번호에 송신을 할 수 있다. 또한 서버가 클라이언트에 알려주어야 하고, 클라이언트가 수동으로 포트 번호를 적어주어야 한다.(**수동 포트 번호가 우선**적으로 선택된다.)

-  윈도우즈 시스템32 - drivers - etc - services 파일에 기본적인 포트 번호가 저장되어 있다! 

- **되도록 규범에 맞춰야 통신이 원활하게 돌아간다.**

- 포트 번호는 충돌 대신 회피가 존재. 한 컴퓨터 내의 두 개의 프로그램이 같은 포트 번호를 사용할 수 없다. 세팅을 막아 충돌을 회피하는 방식

- 2^16 = 0 ~ 65536

### Network Information Center(NIC, 네트워크 정보 센터)
- 인터넷 주소 자원과 정보를 관리하는 기관
- 국제 인터넷 주소 관리 기구(ICANN)에서 운영하며 세계 각국에 네트워크 정보 센터(NIC)가 있음
- 국내에는 한국인터넷진흥원(KISA) 하에 한국인터넷정보센터(KRNIC)이 있음
    - 국내의 IP 주소와 도메인 등록/관리

## 전송 모드
- Unicast(유니캐스트) 1:1
- Broadcast(브로드캐스트) 1:m(불특정다수)
- Multicast 1:n(특정 다수)

### Unicast 전송모드

- 데이터를 전송하려면 주소가 필요하다. ip, mac, port 번호가 각각 송신지와 수신지가 있어 6개의 주소가 필요하다.

- DNS가 문자 주소의 수신지 IP 주소를 알려준다
- ARP가 수신지의 IP 주소로 MAC 주소를 알려준다.

- L4, L3 주소는 수신-송신 구조지만 L2 주소는 송-수 구조이다.

### 브로드캐스트 주소
- 가상주소
- 시스템/애플리케이션 운영 시 사용됨
    - 시스템: 운영체제
- 운영체제 설치 시 할당
- BIP: 255.255.255.255(로컬 브로드캐스트 주소)
    - directed 브로드캐스트 주소: X.X.X.255/24
    - 브로드캐스트는 기본적으로 로컬에서 사용하는 주소
    - 비연속 네트워크(망이 분할된 경우)에는 다이렉트 브로드캐스트 주소를 사용해야 한다. --> 보안상 좋지 못해 내부용으로 제한
- BMAC: FFFF.FFFF.FFFF

- 수신지 주소에 브로드캐스트 주소가 세팅되어 있으면 브로드캐스트 패킷이라고 판단한다.

- 운영체제 종류에 상관없이 같은 주소를 가진다.
    - 따라서 모든 호스트가 받게 된다.

### Multicast 전송모드
- 가상 주소
- 시스템/애플리케이션 운영 시 사용됨
- 특정 시스템/애플리케이션 **설치 시 할당**
- MIP: ip 주소가 224~239 사이로 시작: 224~239.x.x.x
- MMAC: 0100.5EXX.XXXX(제조회사 번호에 0100.5E를 고정)
    - 멀티캐스트는 제조회사 번호를 25비트까지로 본다.
- 특정 IP 주소를 가진 호스트에만 송출한다. 특정 IP를 다수가 지님

- **멀티캐스트 구분은 수신지IP주로로 확인하는 것이 좋다!**
- 예를 들어 아프리카TV를 본다고 할 때, 프로그램을 설치 시 멀티캐스트 주소가 할당된다.

- 255.255.255.255, 224.0.0.2: 보안설정 프로그램을 설치 시 비활성화될 수 있음

- 세팅 순서
    1. 운영체제 설치 시 로컬 브로드캐스트, 멀티캐스드 5개 설치
    2. TCP/IP로 IP 설정
    3. 다이렉트 브로드캐스트 설정

- **송신지 주소에는 브로드캐스트나 멀티캐스트 주소가 올 수 없다!**

- 멀티캐스트는 복잡하고 어려운 구조.
    - 주소 뿐만 아니라 멀티캐스트 프로토콜들과 주소망에 대한 이해가 필요.
    - WAN용 멀티캐스트 프로토콜: 엄청 많이 존재
    - LAN용 멀티캐스트 프로토콜: IGMP

### 정리
- 각 패킷이 어떤 종류의 패킷인지를 구분할 수 있어야 한다.
- 유니캐스트를 제외하면 DNS나 ARP의 도움을 받지 않고 자동 변환이 된다.




###  ARP
- IP 주소에 대응되는 MAC 주소를 조회 변환해 주는 서비스
- ARP 패킷 종류
    1. ARP request 패킷
        - 송신지가 수신지의 MAC 주소를 조회하기 위해 보내는 질의 패킷
        - 브로드캐스트 방식으로 운영 --> 자신에게 해당되지 않으면 패킷 폐기
        - 자기 자신을 명확하게 알려줌(누가 요청한건지)
    2. ARP reply 패킷
        - ARP request에 대해 응답하는 패킷
        - 유니캐스트 방식으로 운영(리퀘스트 패킷에서 수신지 주소를 파악)

- ARP Cache Table
    - ARP 캐쉬 테이블 확인 명령어: arp -a
    - 동적과 정적이 존재. 
        - 정적은 운영체제나 관리자가 필요해 저장해두는 테이블로 전원을 껐다 켜도 복원됨
        - 동적은 ARP request/reply과정을 통해 추가된 테이블

    - 테이블 설정 이유: 반복 작업에 대해 빠른 응답을 하기 위함. 캐시 테이블에 존재하지 않을 경우에 request/reply 과정을 진행

    - aging time: 네트워크 장비에서 특정 정보(ARP 주소, 동적 Cache Table 등)가 얼마 동안 유지될지를 결정하는 시간 설정. 이 시간이 지나면 해당 정보는 삭제되거나 갱신되어야 함. 평균 2~4분

## 계층별 장비
- Switch (L2 --> mac)

- Hub (더미 허브:L1, 스위치 허브: L2)

- Gateway 기능
    - 망과 망을 연결시켜 준다.(내부에서 외부로, 외부에서 내부로의 통신에서 벽 역할을 한다.)
    - 장비: 라우터, **멀티레이어 스위치(자주 사용)**, 게이트웨이(방화벽)

- 네트워크 장비는 컴퓨터 본체에 네트워크 카드가 두 개 꽃혀있다고 생각하면 된다.

- 한 뱡향의 화살표는 허브
- 직선은 LAN 케이블
- 번개모양은 WAN 케이블

- ping 과정

    3. arp reply    (arp 캐시 테이블 등록 작업)
    4. echo-request
    5. echo-reply

    - RTT: 질의에 대한 응답이 올 때까지의 시간.
    - 패킷 로스를 보고 잘 갔는지 확인 가능

    - request time out: 핑을 보낼 때 echo request를 보낼 때 타이며 안에 응답이 오지 않을 경우 출력되는 메시지. 상대방이 다운됐다고 생각 X.

### Forwading과 Flooding
- Forwarding: 하나의 **송신지 포트**(하나의 물리적 인터페이스)에서 하나의 **수신지 포트**로 트래픽 전송(유니캐스트 방식과 유사.)
- Flooding: 송신지 포트를 제외한 나머지 포트들로 트래픽 전송(브로드캐스트 방식과 유사)

- **전송 방식과 비슷할 수 있으나 장비에서는 절대로 유니캐스트, 브로드캐스트 방식이다 라고 하지 않는다!**



### 스위치(Switch, 2계층 장비)
- 스위치의 테이블은 맥주소와 출구 번호로 구성되어 있다.
- 포워딩 과정
    1. 수신지 MAC 주소 확인
    2. MAC 주소 테이블 확인
    3. 존재하면 해당 주소로 전송(정확하게 알려준다.)
- 플러딩 과정
    1. MAC주소 확인
    2. MAC주소 테이블 확인
    3. 존재하지 않으면 모든 목적지에 전송(없어도 전송은 한다 --> 오버헤드 발생 가능)

    ### **둘 다 유니캐스트 패킷인데, 포워딩과 플러딩의 방식으로 전송하는 것이다.**

- 스위치는 2계층 장비
- 경로 데이터베이스 이름: MAC Address Table
- 전송 방법 (유니캐스트 패킷 전송 방법)
    * 포워딩: 목적지 정보(수신지 맥주소)가 경로 데이터베이스에 있는 경우
    * 플러딩: 목적지 정보가 경로 DB에 없는 경우
-  브로드캐스트나 멀티캐스트 패킷: 플러딩
- **스위치는 브로드캐스트나 멀티캐스트에 대한 제어 능력이 없다!**

- **경로 데이터베이스 특징: 브로드캐스트나 멀티캐스트와 같은 주소가 등록되지 않는다!**
    


### Router(3계층 장비)
- 3계층 장비에는 라우터 외에도 멀티레이어 스위치가 존재한다.
- 실제 사내망은 멀티레이어 스위치를 사용하고 있다.

- 라우터는 IP를 인지하는 가장 대중적인 3계층 장비
- 보통 라우터 밑에 스위치가 붙어 있다고 생각해야 한다. --> X.X.X.0 방식으로 테이블 기록
    - MAC address table = MAC ass : 출구 번호
    - 라우팅 테이블 = NetID : 출구 번호
    - arp cache table = IP address : MAC address

- 장비들은 목적지 정보를 먼저 확인한다
    1. 수신지 IP 확인
    2. 라우팅 테이블 확인
    3. 있으면 포워딩 방식으로 전송

    1. 수신지 IP 확인
    2. 라우팅 테이블 확인
    3. 없으면 **폐기**(스위치와의 차이)

- 경로DB에는 브로드캐스트, 멀티캐스트 주소 기록 X
- 멀티캐스트는 따로 테이블이 존재
- 스위치는 브로드캐스트 패킷에 대해 플러딩 방식을 사용하지만, 라우터는 폐기한다.

### Hub(1계층 장비)
- 더미허브를 말하며, 1계층 장비이다.
- 경로 데이터베이스가 없다
- 전송 방법은 플러딩만 사용(조건 없이 모두 전달 --> 오버헤드 발생 가능)
    - 1계층은 주소 정보가 존재하지 않음
    - 전기 전자 형태로 전달. 기준 주소가 없어 데이터가 들어오면 플러딩 방식으로 내보낸다.(플러딩 방식만 사용)



### 정리
- 허브(하급): 플러딩
- 스위치(중급): 플러딩, 포워딩
- 라우터(상급): 포워딩
- 상급으로 갈 수록 필터 기능이 높아진다.

- 스위치는 테이블을 스스로 관리하는 반면, 라우터는 라우팅 테이블을 관리자가 관리할 수 있다.

    ### 외부망으로 ping 요청을 했을 때 정상적인 통신이 안되는 이유
    - 라우터를 통해 다른 망으로 핑을 보낼 때 요청시간 오버로 인해 echo request는 도달했지만 3.0망의 라우팅 테이블에는 1.0망이 없어 폐기해버린다.
    - 1.10에서 5.10으로 통신이 안되는 이유는 게이트웨이1에 5.0망에 대한 정보가 없기 때문이다.
    - request를 보내도 reply를 받지 못하면 통신할 수 없다! 돌아오는 정보가 라우터에 없으면 폐기해버린다.

    - 첫 패킷은 길을 닦는 희생양 취급. 조회 후 캐쉬에 올라가기 때문에 첫 패킷은 로스될 수 있다.

    - ping을 두 번 하면 첫 ping 패킷에 이어서 보낸다고 생각해야한다.(8개를 보내는 것)

    - tracert 시 시간의 의미: 첫 번째 echo-request에 대한 RTT, 두 번째 echo-request에 대한 RTT, 세 번쨰...
        - 통신 속도가 느려지는 구간을 찾을 수 있다.

## 제일 중요
### Media Translation
- 3계층 이상의 장비에서 처리
- 패킷이 출발지에서 목적지까지 가는 동안 3계층 장비를 거칠 때마다 L2헤더 변경
- 3계층 주소(IP) 변환 없음
- 2계층 주소는 스위칭 환경에 따라 변환

- 패킷을 보낼 때 목적지 MAC 주소에 실제 MAC주소가 들어가지 않는다! (보낼 때 상대의 MAC 주소를 모르는 상태)
    - arp를 먼저 보내면 되지 않을까? --> ARP request는 라우터 밖을 나갈 수 없다.(브로드캐스트 방식이기 때문. 라우터가 폐기)
    - 따라서, 외부망으로 보낼 때는 **목적지 MAC 주소 부분에 게이트웨이의 MAC 주소**를 넣는다.

- 강의 자료 30p
    - 패킷 경로             SIP   DIP      SMAC   DMAC
        1. PC1 --> RA      21.7  34.250   6666   2222
        2. RA --> RB       21.7  34.250   9999   3333   
        3. RB --> FTP      21.7  34.250   7777   5555 

    - 3계층 주소(IP)는 바뀌지 않는다.

    ### Media Translation이란
    - **송신지에서 목적지까지 가는 동안 IP는 바뀌지 않지만 2계층 주소가 계속 바뀌는 것.**

    - LAN은 2계층으로 정해진 주소가 없다. LAN카드마다 주소가 다름
        - 따라서, LAN을 거칠 때마다 2계층 주소가 바뀌는 것을 확인할 수 있다.

    - **수신지 IP와 수신지MAC은 싱크가 되는 것이 아니다!!**
    - 공백만 아니면 된다.
    - 외부망의 경우 둘이 같은 주소인지 절대 알 수 없다.
    - MAC 주소는 2계층 장비, IP 주소는 3계층 장비를 위함

- 송신지 IP와 수신지 MAC이 다를 때 발생 가능한 현상
    1. 응답이 원래 송신자에게 도달하지 않음
        - 수신자는 IP헤더의 송신지 IP를 보도 응답을 보내지만, 이 응답은 MAC 주소를 기반으로 전달되어 원래 송신자에게 전달되지 못한다.
        - MAC 주소가 실제 송신자와 다르면 응답이 다른 장치로 전달되거나 유실될 수 있다.
    2. ARP 캐시 충돌 또는 오류
        - 수신 측은 ARP를 통해 송신지 IP에 대응하는 MAC 주소를 학습
        - 이때 MAC 주소가 실제 송신자와 다르면, ARP 캐시에 잘못된 매핑이 저장되어 이후 통신에 문제가 생길 수 있다.
    3. 보안 경고 또는 패킷 드롭
        - 일부 네트워크 장비나 보안 시스템은 IP-MAC 불일치를 감지하면 해당 패킷을 드롭하거나 로그에 기록함 --> ARP 스푸핑 또는 MAC 위조 공격으로 간주될 수 있음
    4. 라우팅 오류
        - 라우터는 IP 기반으로 경로를 결정하지만, 실제 패킷 전달은 MAC 주소를 기반으로 이뤄져요
        - MAC 주소가 잘못되면 패킷이 잘못된 경로로 전달되거나 응답이 돌아오지 않을 수 있다.

### DNS
- arp cache table이 존재하면 dns cache table도 존재한다
- nslookup www.naver.com: 실제로 네트워크 주소를 알려주지 않고, 접근 가능한 가상 주소를 알려준다. 이 주소를 알려주는 dns 서버의 정보를 알 수 있다.
- ping 문자 주소를 할 경우 주소에 대한 ip를 알아내어 이 주소로 ping을 보낸다. 실제로 응답한 서버의 ip가 노출될 가능성이 있어 ping은 닫아두는 경우가 많다.
- ipconfig /flushdns를 통해 캐쉬 테이블 삭제가 가능하다.

- 문자 주소와 ip로 구성된 **테이블을 DNS 캐쉬 테이블**이라고 한다.

- **hosts 파일**에는 **IP주소 : FQDN** 형태로 저장되어 서버의 역할을 한다.
- dns에 ip주소를 묻고, 받은 ip 주소를 통해 통신한다.
- dns는 외부망에 있는 서버를 더 많이 사용한다.
- 한 번 방문한 경로는 dns의 cache table에 기록하여 다음 방문 시에 더 빠르게 접속할 수 있다.
