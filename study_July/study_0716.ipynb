# 사내망 구성도
## 보안 솔루션
### HIDS(Host-based IDS)
- **서버에 직접 설치**됨에 따라 네트워크 환경과 무관
- **호스트의 자원 사용 실태, 로그 등을 분석하여 침입 여부 탐지**    
- **무결성 체크** 기능이 주요 기능
    - 무결성 점검에 의해 침입여부 식별
    - 최초 설치 시 초기 데이타베이스에 중요 파일들에 대한 **해시값 저장**
    - 주기적으로 중요 파일의 해시값 변조 유무를 검사/탐지/분석하여 결과 보고
- 오픈 소스 IDS : Tripwire(트립와이어)




### NIDS (Network-based IDS) 

- 망 내에 떠돌아다니는 트래픽들을 검사
    - **패킷을 수집하는 기능**이 필수

- 네트워크 상에서 일어나는 **침입 시도를 탐지**
- **네트워크 세그먼트 당 하나의 장비만 설치**하면 되므로 설치 용이
    - 네트워크 세그먼트: 특정 구간. 관제 범위

- 패킷 수집을 위해 **mirroring** 기능 이용(Sniffing 환경 필요)
    - Mirroring : **패킷을 복사한 다음 복사한 패킷을 NIDS 장비로 전달**
    - **스위치**의 미러링 포트를 이용하거나 **TAP 장비**를 통해 패킷 복사

- 수집된 패킷은 분석을 위해 **필터링과 축약과정**(reduction)이 필요
    - 필터링은 불요한 정보를 제거하는 과정으로 지정된 수준의 데이터만 수집
    - 축약은 통계적/수학적 기법을 적용하여 반복되는 데이터를 줄이는 과정

- 오픈 소스 NIDS : Snort

### False Negative & False Positive

- 미탐지 (False Negative)
    - 공격을 **탐지하지 못하는 경우**
    - 시그니처 기반의 탐지 시스템의 경우 미탐지가 높음

- 오탐지 (False Positive)
    - 공격이 **아닌 것**을 공격으로 탐지하는 경우
    - 행동 기반의 탐지 시스템의 경우 오탐지가 높음

### IDS 탐지 방법

- 오용탐지(Misuse Detection)
    - **시그니처 기반의 탐지**

    - 알려진 공격법이나 보안 정책에 위반하는 행동에 대한 패턴 탐지

    - 공격 분석 결과를 바탕으로 패턴 설정

    - 패턴(시그니처)과 비교하여 일치하는 경우 불법 침입으로 간주

    - 공격 패킷마다 고유한 시그니처가 존재. 이를 통해 공격 패턴을 찾아낸다.
        - 예시: 수업 중 자는 것을 공격 시그니처로 잡아둔 경우 자는 행동은 모두 공격으로 간주됨

    - 굉장히 오랜 기간에 걸쳐 공격 패킷에 대한 시그니처를 정의한다.
        - 장시간 패킷을 분석하여 패턴을 설정

    - 실습은 와이어샤크를 통해 공격 패킷의 공통점(시그니처)을 찾고, snort를 이용하여 탐지 룰을 생성

- 이상탐지(Anomaly Detection)
    - **행동 기반의 탐지**
    - 값, 수치를 정해두어 똑같은 시그니처에 대해서도 **일정 수치를 넘어가야** 공격으로 간주한다.(유연한 방법)

    - 정상범위를 벗어나는 데이터를 탐지하는 방법

    - 정량적인 분석, 통계적 분석을 사용

    - 형태 관찰, 프로파일 생성, 프로파일 기반으로 이상 여부를 확인

    - I/O 사용량, 로그인 횟수, 패킷양 등

    - 룰을 생성하는 것이 아니라 **평상시의 형태**를 관찰하여 기록하고, 이상치에 대한 **로그**를 남겨 관리자에게 알린다! 
        - AI를 활용할 수도 있다
        - 자동화 성향


### 탐지 방법의 장단점
- 오용탐지 (Misuse Detection) 
    - 장점
        - 오탐률 (false positive)이 낮음
        - 트로이 목마, 백도어 공격 탐지 가능
    
    - 단점
        - 미탐률 (false negative)이 높음
        - 새로운 공격 패턴 갱신 필요
        - 패턴에 없는 새로운 공격에 대해서는 탐지 불가능

- 이상탐지 (Anomaly Detection)
    - 장점
        - 미탐률 (false negative)이 낮음
        - 인공지능 알고리즘 사용으로 스스로 판단하여 수작업의 패턴 업데이트 불필요
        - 알려지지 않는 새로운 공격 탐지 가능
    
    - 단점
        - 오탐률 (false positive)이 높음
        - 정상과 비정상 구분을 위한 임계치 설정이 어려움




### 침입 탐지 시스템의 한계
- 오탐지(false positive)와 미탐지(false negative) 문제 발생
    - 공격에 대한 패턴을 모르다면 분석과 탐지가 어려움
- **실시간 공격을 막을 수 없음**
    - 구성에 따라 다를 수 있다!
    - 시험에선 못막는다가 맞다!
- **단편화(fragmentation), 난독화, 암호화**와 같은 기술은 감지하지 어려움


### 침입 방지 시스템 (Intrusion Prevention System)
- 방화벽과 IDS의 단점을 보완하기 위해 나온 장비
    - 방화벽: 제어(차단 기능), 악성코드 공격에 약함(능동형)
    - IDS: 악성코드 탐지률이 높음, 제어기능 약함(수동형)
    - IPS = 방화벽(제어,차단 기능) + IDS(악성코드 탐지)(능동형)

- 침입탐지시스템의 detection 기능과 방화벽의 차단(blocking) 기능 결합

- 이상 행위 탐지(anomaly detection)를 통해 알려지지 않은 공격 패턴에 대응
    - 딥러닝 기반의 탐지가 가능

- 공격에 대한 사전 방지를 조치하는 것으로 in-line 방식으로 설치 및 운영

- 실시간 침입차단, 인터넷 웜, 악성 코드 및 해킹에 기인한 유해 트래픽 차단

- **능동형 보안 솔루션**
    - IDS : 탐지 후 사후에 조치를 취하는 기술. **수동형 보안 솔루션**
    - IPS : 예방적이고 사전에 조치를 취하는 기술

- IPS 필터
    - 필요한 필터만 올려 사용하면 된다!
    - 대량의 데이터가 들어와도 처리할 수 있는 프로세서들을 개발하여 문제 없이 작동할 수 있다!

    - 방화벽 필터: 액세스 제어 및 패킷 필터링 기능 (IP, 포트 등 패킷 필드별로 막을 패킷 정의)

    - 트래픽 모디터링 및 QoS 필터: 프로토콜별, 서비스별, IP 영역별 QoS 기능 제공

    - 프로토콜 무결성 확인 필터: FTP, DNS, 메일, 웹 등 TCP/IP 프로토콜 동작 표준에 위반하는 패킷 조사

    - 시그니처 이상 감지 필터: 악성 코드, 취약성, 웜에 대한 탐지 및 차단

    - DoS/DDoS 스캔 필터: DoS/DDoS 스캔

    - L7필터: TCP /IP 단편화, 웹 우회 공격 등 L7 프로토콜 디코딩을 통한 필터

    - 데이터 컨텐츠 내용을 기준으로 필터링 여부 결정

- 외부 네트워크에서 방화벽으로 들어오기 전에 설치하여 필터링 진행.
    - 외부 네트워크에서 들어오는 패킷들을 모두 검사하면 속도 저하가 발생할 수 있다.



### Web Application Firewall(WAF) 
- 웹 컨텐츠(HTTP/HTTPS) 를 분석하여 공격을 탐지 및 차단하는 기능을 가진 방화벽
    - OWASP Top 10의 웹 공격에 사용되는 요청 파라미터 패턴을 분석
    - 웹서버를 통해 데이터베이스를 공격하는 유형이 많이 늘어나면서, 웹서버 앞에 웹공격을 막는 WAF를 설치하는 경우가 많아졌다.

- WAF 기능
    - 서버의 정보와 개인 정보 등의 중요한 정보 유출 차단 (service out) 
    - 정보유출 방지, 부정 로그인 방지, 웹 사이트 위변조 방지등에 활용

    - 정보 유출 방지: 개인 정보가 웹 게시판에 게시되거나 개인정보가 포함된 파일등이 웹을 통해 업로드 및 다운로드 되는 경우에 대해 탐지하고 대응

    - 부정 로그인 방지: 무작위 대입 등의 방식으로 웹사이트 로드인을 시도하는 경우와 같이 비정상적인 접근에 대한 제어

    - 웹사이트 위변조 방지: 해킹 후 웹사이트를 위변조 하는 경우 이를 탐지하고 대응

### **NAC(Network Access Control)**
- 보안쪽에서 중요!!

- 재택 근무가 많아지면서 외부에서 서버로 접근하는 경우, 외부 접속자의 PC가 악성 코드에 감염됬거나, 보안패치 설치가 안되어 있는 경우 문제가 발생할 수 있다.
- 이를 예방하기 위해 검사하는 기능!
    - 점검항목
        - 정상적인 장치/사용자 인증 확인(노트북의 시리얼/MAC주소, 사원 번호 등)
        - 안티 바이러스 설치 여부 점검
        - 보안패치 설치 여부 점검
        - **위 항목 중 하나라도 위반 시 접근 제어**



### VPN (Virtual Private Network)
- 인터넷과 WAN
    - 인터넷의 일부분이 WAN. 전 세계의 WAN을 모아 만든 거대한 망이 인터넷이다.
    - 본사와 지사를 연결하기 위해 통신사에게 요청하여 WAN 회선을 구축할 수 있다. 이 경우 통신 중 보안 사고가 발생하면 통신사가 책임을 진다. 그러나, 인터넷을 이용할 경우 보안 사고가 발생해도 책임자가 없다.
    - 비용 문제로 인해 인터넷을 이용해서 연결할 수도 있다.
    - WAN에 비해 인터넷 망이 특히 느리지도 않다.
        - 보안 문제!!
        - 안전성이 없다.
    
- 인터넷 망을 이용할 때 **안전성을 보장**해주기 위해 등장한 기술이 VPN이다.
    - 공용 망을 개인 망처럼 느껴지게 하는 기술
    - 암호화 기술 사용

- 계층별 암호화 프로토콜과 VPN
    - VPN: 캡슐화 기술(암호화 헤더를 추가시킴)
        - 헤더: 제어 정보

    - 2계층: L2TP, PPTP, VPN -->  L2TP VPN, PPTP VPN
        - PPTP: MS 2L 암호화 프로토콜
        - L2F: Cisco L2 암호화 프로토콜
        - PPTP + L2F --> L2TP

    - 3계층: IPSec, VPN --> IPSec VPN
        - 망대 망으로 구성할 경우 주로 사용
        - 안정성이 매우 높음

    - 4계층: TLS, VPN --> SSL VPN
        - 개인대 개인간 통신에 주로 사용
    - 5계층: SSL, VPN --> SSL VPN
        - 개인대 개인간 통신에 주로 사용
    - 7계층: HTTPS, SSH etx

- L3 이상의 장비는 대부분 VPN 설정 가능.

- VPN = 캡슐화 기술 --> 터널링 기술
- 세팅은 쉽지만, 운용이 어렵다(이론)

### UTM
- 규모가 작은 네트워크에서는 통합 장비인 UTM을 주로 사용한다.

- 다양한 보안 기능을 하나의 장비로 통합하여 제공하는 보안 솔루션
- 네트워크는 간단해져서 투자 비용은 줄고 관리는 쉬워짐
- **중소 규모의 네트워크**를 중심으로 많이 적용되고 있는 장비

- 다양한 장비의 기능들을 하나로 때려넣은 장비

- 장비 가격은 기능도 중요하지만, 처리율이 중요하다!
    - 대규모 데이터에 대한 처리가 잘 되면 가격이 높아진다.

## 보안망 구성도
-  방화벽을 기준으로 3가지 망, 또는 4가지 망으로 구분
- 방화벽에는 여러 개의 LAN카드를 장착할 수 있다!
- 망 구분하는 연습하기

- **InLine** 방식으로 보안장비를 설치한다면?
    - 회선에 직접 설치.
    - IPS 자리에 IDS도 설치 가능하다. (약하지만 차단 기능 활성화)
    - 모든 패킷을 분석하므로 **실시간 방어**가 가능하다.

- **Out-of Path** 방식(미러방식)
    - 경로 밖에 설치하여 패킷을 복사하여 분석하기 때문에 실시간 방어가 불가능하다.

### InLine 모드
- 물리적 네트워크 경로 상에 보안장비를 설치
- 네트워크을 통과하는 모든 트래픽들이 보안장비를 거쳐 가도록 하는 모드
- 패킷 차단 목적의 장비에 적용 (예) Anti-DDoS, Firewall, IPS 등
- 장점 : 실시간 패킷을 탐지하고 차단
- 단점 : 장비에 장애가 발생 할 경우 전체 네트워크 장애로 확산 될 위험성
    - 전체 네트워크 가용성에 영향을 줄 수 있음

### Out of Path(Mirror) 모드
- 미러링 장비(예. TAP)를 통해 복제된 패킷을 받아서 탐지하는 모드
- 패킷 차단 기능이 없는 탐지 목적의 장비에 적용
    - IDS, Anti-APT, Network Forensic 등(네트워크 포렌식은 패킷을 확인하는 것 자체를 의미. 디지털 포렌식과는 다른 것)
- 네트워크 경로를 벗어난 곳에 위치
- 장점 : 전체 네트워크 가용성에 영향을 주지 않으면서 패킷을 탐지할 수 있음
- 단점 : 복제(복사) 된 패킷을 탐지하기 때문에 실시간 패킷을 차단하기 어려움


### DMZ 서버
- 인트라넷 서버는 직원들을 위한 서버, Storage나 DB와 같은 업무용 서버를 설치
- DMZ에는 내부 사용자와 외부 사용자가 모두 접근할 수 있는 서버를 설치
    - Public Server
    - 웹 서버, 메일 서버 등을 설치
    - 각 서버에 맞는 보안 장비를 앞에 설치

## Port Scan
- 실제 공격 방법을 결정하거나 공격에 이용될 수 있는 네트워크 구조, 시스템이 제공하는 서비스 등의 정보를 얻기 위해 수행되는 방법
    - 공격 대상 보안 장비 사용현황
    -  우회 가능 네트워크 구조
    - 시스템 플랫폼 형태
    - 시스템 운영체제의 커널 버전의 종류
    - 제공 서비스 종류

    - 공격할 수 있는, 대상 시스템의 포트가 열려있는지 확인하거나 방화벽을 확인

- nmap(network mapper)
    - 스캔 도구
    - TCP 3WHS 사용
    - 운영체제 종류 및 사용 서비스에 대한 정보 스캔도구
    - 와이어샤크에 기본적으로 설치되어 있음
    - 포트가 몇 개 열려 있는지 확인
    - 대상 MAC 주소 확인
    - nmap (옵션) 192.168.10.20
    - -sT -O 옵션: MAC주소, 홉 수, 운영체제 정보 획득 가능
    - **SYN 패킷**이 많은 것을 확인할 수 있다.
    - (filter: tcp.flags.syn == 1 && tcp.flags.ack == 0)


### TCP Full Open Scan(자신의 정보 노출)
- **포트가 열려 있는 경우 SYN/ACK 패킷 수신** --> 이 포트들을 이용
- 닫힌 경우는 해당 포트가 RST를 보냄
- 차단 정책으로 차단되어 있는 경우는 응답이 돌아오지 않는다.
    - SYN에 대한 응답이 안오는 경우 방화벽에 차단된 것으로 판단.

- SYN/ACK에 ACK 패킷을 전송함으로써 연결을 완료
- 스캔하고자 하는 포트에 접속을 시도해 **완전한 TCP 연결**을 맺어 신뢰성 있는 결과 얻음
- **속도가 느리고 로그를 남기므로 탐지가 가능**하다는 단점을 가짐

### TCP Half Open Scan
- 세션에 대한 로그가 남는 TCP Full Openscan을 보안하기 위한 기법

- 공격대상으로부터 SYN/ACK 패킷을 받으면 공격자는 RST 패킷을 보내 연결을 끊음
- --> 세션을 완전히 연결하지 않음

- RST 패킷을 보낸다면 그 포트가 열려있음을 의미.

- **로그를 남기지 않아 추적이 불가능하도록 하는 기법**

### 스텔스 스캔 (Stealth Scan)
- 3Way Handshaking 연결 기법을 이용한 것이 아님
- **TCP 헤더를 조작**하여 특수한 패킷을 만들어 스캔 대상의 시스템에 보내어 그 응답으로 포트 활성화 여부를 알아내는 기법

- 세션을 성립하지 않고 공격 대상 시스템 포트 활성화 여부를 알아내기 때문에 공격 대상 시스템에 로그를 남기지 않음

- 공격 대상의 시스템 관리자는 어떤 IP를 가진 공격자가 시스템을 스캔 했는지 확인 할 수 없음

- nmap -sF[대상IP] == FIN scan
    - TCP는 포트가 열려 있는 경우에, 세션을 맺지 않은 공격자가 FIN을 보내면 무시한다.
    - 그러나, 닫힌 포트는 해당 FIN에 RST 응답을 보내는데, 이를 이용한 공격 방법이다.
        - RST,ACK가 오면 닫힌 포트, 응답이 오지 않으면 열린 포트 혹은 차단되어 있는 포트

    - 차단되어 있는 경우에도 무시



### Xmas Scan
- nmap -sX [대상IP]
    - TCP 헤더 내에서 UGR, PSH, FIN을 동시에 설정해서 전송
    - 포트가 열려 있는 경우 응답이 없음

    - FIN, PSH, URG 플래그 패킷을 전송하여 응답이 없는 패킷을 열린 포트로 판단(혹은 차단된 포트)
    - 말도 안되는 플래그 설정. PSH,URG는 전송 중, FIN은 연결 종료 시 보내는 플래그.

    - 시그니처 == FIN, PSH, URG

### Null scan
- $nmap –sN [대상IP]
    - TCP 헤더 내에 플래그 값을 설정하지 않고 패킷을 전송
    - 닫힌 경우 RST/ACK 신호를 응답함
    - 포트가 열려 있는 경우 응답 없음

    - 시그니처 == null flag([<None>])

 ### UDP Scan
- UDP는 3-way handshake 와 같은 절차가 없음
- UDP 패킷을 전송 시 **열려 있는 포트로부터 특정 UDP 응답값으로 수신**
- 수신측의 포트가 닫혀 있는 경우 ICMP Port Unreachable 에러 메시지를 통해 포트 활성화 유무 확인
- 포트가 열려 있으면 UDP 응답이 돌아온다.
- 포트가 닫혀 있으면 3계층의 도움을 받아 ICMP메시지로 닫혀있다는 응답 반환
    - ICMP type 3, code 3 message == port unreachable(닫혀있음)
- $nmap –sU [대상서버IP]

- 시그니처 == 단기간에 많은 destination unreachable (port unreachable) 메시지

## Pharming Attack
- TCP는 닫혀 있는 포트에 너무 친절하다.
- **포트 스캔만으로도 감옥에 갈 수 있다!!** 함부로 시도하지 말 것

- pharming
- 피싱(Phishing) + 조작(Farming)의 합성어

- 정상 사이트에 접속하더라도 가짜 사이트로 접속을 유도하여 금융거래 정보를 빼낸 후 금전적인 피해를 입히는 사기 수법

- 위조 사이트를 만들어둔 후 접속 유도를 통해 정보를 빼내는 사기 수법
    - 정상 사이트로 위장한 위조 사이트로 접속 유도 --> 피싱
    - **파밍 --> 정상 사이트를 직접 입력해도 위조 사이트로 접속하도록 하는 것**
        - ARP Spoofing, DNS Spoofing을 사용.
        
    - 이를 막기 위해 https로 넘어감. 위조 사이트가 아님을 증명할 수 있음


- Spoofing
    - '속이다'는 의미
    - 인터넷이나 로컬에서 존재하는 모든 연결에 spoofing 가능
    - 정보를 얻어내기 위한 중간 단계의 기술로 사용하는 것 외에 시스템을 마비 시키는 데 사용할 수도 있음
    - 수신지 MAC, IP 주소를 속이는 행위
        - IP: DNS 스푸핑 이용
        - MAC: ARP 스푸핑 이용

- ARP Spoofing
- DNS Spoofing




### ARP Spoofing
- MAC 주소를 속이는 것
- 2계층에서 작동해 공격 대상이 같은 랜에 있어야 함

- 타깃이 GW에 ARP request를 보낼 때 브로드캐스트로 보내기 때문에 공격자가 먼저 MAC주소를 속여서 알려줄 수 있다.
- ARP Cache Table에 매치되지 않은 정보가 등록이 된다. 
    - 공격자가 지속적인 reply 패킷을 보내 테이블을 유지한다.(aging time 고려)

- 타깃이 외부망을 이용할 때 공격자에게 트래픽이 전송되어 도청당할 수 있다.
    - 공격자는 패킷을 받아 게이트웨이로 리다이렉션을 해주면 된다.


- 실습 
    - 192.168.10.10*kali의 MAC주소: ~3b51
    - 게이트웨이 MAC: 1393(정상적인 주소)

    - arpspoof -i eth0 -t 192.168.10.40 192.168.10.2
    - 윈도우(10.40)이 게이트웨이에게 주소를 물어보면 자신의 주소를 알려주는 작업.
    - 지속적으로 reply를 보낸다.
    - 192.168.10.40의 ARP Cache Table에서 게이트웨이의 MAC주소가 kali의 MAC주소로 변경되어 있는 것을 확인할 수 있다.

    - IT 기술을 몰라도 할 수 있는 공격 방법.
    - setoolkit

### DNS Spoofing
- 실제 DNS 서버보다 **빠르게** 위조된 DNS response 패킷을 보내 공격 대상이 잘못된 IP 주소로 웹 접속을 하도록 만드는 공격 방법

- 클라이언트는 이미 DNS response를 받았으므로 **정상 DNS response는 drop**
    - 늦게 온 패킷은 drop
- 보통 DNS 서버는 외부망에 설치 --> 내부망보다 느릴 가능성이 높다!
- 가짜 DNS 서버를 설치하는 방식도 존재

- **수신지 IP주소를 속이는 것이 DNS 스푸핑이다.**


- 실습
    - ARP,DNS Spoofing을 이용하여 www.naver.com을 입력했을 때 www.facebook.com 페이지로 이동하도록 파밍
    - nano dns --> dns 파일 생성 및 열기
    - 192.168.10.10 www.naver.com --> 주소 등록 
    - dnsspoof -f /dns --> DNS Spoofing
    - 윈도우에서 arp -a, nslookup www.naver.com 명령어를 통해 속은 상태인 것을 확인할 수 있음
    - explorer에서 www.naver.com을 입력하면 위조 사이트인 www.facebook.com으로 이동되는 것을 확인할 수 있다.
