# TCP/IP 프로토콜
## OSI 7 layer 참조 모델(Open System Interconnection)


- 개방형 시스템: 원격 접속이 가능하게 해둔 시스템
    - 랜카드 장착, 파일 공유, 케이블 연결 등의 작업이 되있는 경우
    - 양측이 모두 시스템을 구축해두어야 한다.
    - 두 개방형 시스템간에 통신을 허용하기 위해 7가지 기능이 필요했다 -> OSI 7 Layer
    - 반대는 closed System(폐쇄형 시스템)
        - 외부와 연결되지 않거나, 매우 제한된 방식으로만 작동하는 시스템
        - 외부 API를 허용하지 않는 소프트웨어
        - 인터넷 없이 작동하는 로컬 네트워크 프로그램(군대 인트라넷 등)

- **각 계층의 고유 기능**

### OSI 7 참조 모델

- 상위 3계층을 Software 기능 (개발자, 빅데이터 분석, 데이터베이스..)
- 전송 계층을 Middleware 기능 (server -> system 보안)
- 하위 3계층을 하드웨어 기능 (Infra)


### Application Layer
- 사용자에게 다양한 User Interface(UI) 제공
- UI는 어떠한 서비스를 받기 위해 사용자 입력을 받는 툴이다.
- UI를 통해 데이터 생성
- 요약: **특정 서비스에 대한 UI 제공, UI를 통해 데이터 생성**

### Presentation Layer
- 변환 작업을 하는 계층
- Code 변환(부호화, encoding(이진화))
    - 인코딩 예: 아스키, 유니코드, BCD 등
- 압축화 (예): jpg, png, bmp, mp4, zip 등
- 암호화 (예): RSA, DES 등

- **이진화, 암호화, 난독화, 압축화**

### Session Layer
- Interhost Communication
- Application 사이의 대화 Start, Maintain, Stop

- **양단간에 서비스 개시(설정), 유지, 해지**
    - 서비스 개시: 인증
    - 서비스 유지: 지속적 인증 작업
    - 서비스 해지: 인증값 만료(보안을 위해. 보통 10분 정도 유효)
    - 서비스를 제공할건지, 유지할건지, 해지할건지를 지속적으로 판단하는 계층
    - 인증값(세션ID) == ID + PW
        - 세션 하이제킹
    - SSL == 인증서 기반 암호화(현재는 사용 X)

- *지속적 인증*
    - http --> stateless. 요청에 대한 작업이 끝나면 세션이 종료.
        - 세션이 생길 때마다 다시 인증 작업이 필요. 이전의 인증 작업 소실
        - 예: 1탭에서 로그인하여 작업. 2페이지에서 새로운 로그인 필요(지속적 인증 X)

    - **인증값**: 클라이언트가 서버에 아이디와 패스워드를 입력하면 서버의 5계층에서 인증값(쿠키 아이디, 세션 아이디) 생성. 그 후 클라이언트에게 인증값을 넘겨주면, 클라이언트는 다음 작업(새로운 페이지) 시 클라이언트의 5계층에서 인증값을 넘겨준다.



### Transport Layer
- 원하는 서비스를 받을 수 있는지 확인해주는 계층
- **양단간(end-to-end)에 가상 회선을 설정, 유지, 해지하는 역할. 서로간의 상태 확인**
- host == end point(이전에는 host라고 불렀지만, 요즘은 end point라고 불린다. 보통 호스트가 엔드 포인트가 됨)
- TCP 프로토콜
    - 3Way handshake

    - Virtual Line(가상 회선) 수립: 
        - 서로의 상태 확인 작업이 끝난 상태
        - 서버 활성화, 서비스 활성화 상태
    - 가상 회선 해지:
        - 서버 비활성화, 서비스 비활성화
    
    - **데이터 송수신 전에 서로간의 상태를 확인하는 작업**


### Network Layer
- 라우팅을 담당하는 역할. 경로 제어 담당
- Router: 경로를 제어하는 기기(장비)
- Routing: 최적의 경로를 찾는 것(선정 작업, 과정)
- Route: 최적 경로

### Data Link Layer
- 2계층 기본 기능: Forwarding(전달 기능)


- LAN은 2계층 구간. 스위치나 허브만 있으면 되는데, 스위치가 2계층 장비이다.
    - 갈림길에서 경로를 결정해주는 장비는 라우터이며, 라우터만 경로 제어를 담당한다.
    - LAN 장비 == 스위치, 허브
    - 공유 장비를 이용해서 데이터를 전송하는 특징이 존재. 
        -  충돌 문제 발생 --> 전송 실패 --> 재전송 --> 전송 지연 문제
    - 이를 해결하기 위한 충돌 제어 기술(**공유 매체 접근 제어 기술**): **CSMA/CD**(충돌 감지. CSMA/CD: IEEE 802.3)
    - 충돌 감지 기술은 유저가 조심해야 함 --> 컴퓨터 유선 LAN카드에 탑재 == 유선랜카드 == 이더넷카드
        - 모든 호스트가 이더넷카드를 장착한 LAN망이 이더넷망이 된다.

- 공유 매체 접근 제어
- 오류 제어 기능
    - 수신자 측에서는 무결성 점검을 해야함
    - 전송되는 과정에서 데이터 오염도 검사
    - 오류 검출 코드를 첨부해서 보냄
        - 주요 오류 검출 기술
            1. 패리티 체크
            2. 체크섬
            3. CRC, 순환 중복 검사
            4. 해밍 코드
            5. 2차원 패리티 검사

- 2계층에서는 하는 일이 많다. 
    - 대표적인 3가지
        1. 데이터 전달
        2. 공유매체 접근 제어
        3. 오류 검출

- 데이터링크 계층은 패킷에서 단일 또는 두 가지의 주소 체계가 나타날 수 있다.
    - LLC: 어떤 프로토콜에게 전달할지
    - MAC: 누구에게 보내야 하는지



- 스위치는 데이터 전달 장치
- 라우터는 최적 경로 설정 장비(2계층 장비 X)


- PBL 과제: CSMA/CD, CSMA/CA, Token Passing, Aloha, CSMA에 대한 정리. 특히, CSMA 3가지에 대한 차이점 알아보기.

### Physical Layer

- **물리적인 연결**을 담당
    - 상위 계층에서의 작업을 물리적으로 전송하는 계층
    - 물리 계층은 데이터를 어떻게 보낼지에 집중
    - 상위 계층들은 무엇을 보낼지에 집중
- 케이블 하나하나의 규약을 정하는 계층으로, 데이터 전송에 문제가 없도록 하는 계층
- 가장 규약이 많은 계층
- 전기적, 기계적, 기능적

### 컴퓨터는 몇 계층일까?
- 각 계층은 하위 계층이 단단하게 지지해주는 건물과 같다.
- 예를 들어 라우터는 3계층이지만 3계층 역할만 하는 것이 아니라 하위 계층의 기능들을 모두 포함한다.
    - 라우터는 3계층 장비이지만 2계층의 스위치로 사용할 수도 있다.

- 따라서, 컴퓨터는 애플리케이션을 다루기 때문에 7계층 하위의 기능이 모두 있어야 온전하게 돌아간다. --> 7계층

- payload: 계층을 내려가는 행위
- 페이로드에 제어 정보가 부착되는 것을 캡슐화
    - 페이로드에 제어 헤더가 붙어 내려가면, 이전 계층의 헤더를 포함한 부분이 페이로드가 된다.(데이터 --> 데이터 + TCP헤더 --> 데이터 + TCP헤더 + IP헤더 --> 데이터 + TCP헤더(포트번호) + IP헤더 + MAC(이더넷)헤더)

## OSI Reference Model과 TCP/IP
- OSI는 국제 표준으로 지정된 참조 모델.
    - IEEE
- TCP/IP 모델은 구현 모델로 현업에서 적용되는 모델
    - 미 국방성에서 군사적 목적으로 개발되어 연구소로 넘어감
    - 가장 안정적인 비상업적 통신 모델
    - 독점을 막기 위해 인터넷 관련 종사자는 비상업적 통신 모델을 이용하여야 하기 때문에 유명하다.

- WAN과 인터넷의 차이
    - WAN은 여러 LAN을 연결한 광역 네트워크
    - 인터넷은 전 세계의 WAN과 LAN을 연결한 거대한 네트워크
    - 따라서, WAN은 인터넷의 구성 요소 중 하나

- OSI 모델에서는 애플리케이션 계층, 표현 계층, 세션 계층 3가지를 켜야 하는 걸 TCP/IP 모델에서는 애플리케이션 계층에 통합시켜 한 번에 실행하도록 구현했다.
    

### TCP/IP Protocol Stack
- 프로토콜 스택: 각 계층을 지원해주는 프로토콜들을 쌓아둔 것

- Application Layer: Servuce. TCP/UDP 방식
    - 예제
        - TCP 방식 수업: **항상 클라이언트가 요청**. 학생이 강사에게 준비가 되었는지 확인. 강사도 학생에게 준비가 되었는지 확인. 확인 완료를 통한 연결 확립. 항상 **요청에 대한 확인 응답**이 존재. 대답이 없으면 다시 응답을 보냄.
            - 장점: TCP는 신뢰도를 보장해줄 수 있다.
            - 단점: 속도가 느리다.

        - UDP 수업 방식: 강사가 학생이 준비가 되었는지 확인하지 않고 일정에 맞게 강의를 한 후 종료. --> 클라이언트가 준비가 되었는지 확인 없이 데이터를 보냄. 다 보내면 끝냄.
            - 장점: 속도가 빠르다
            - 단점: 신뢰도가 없다.

    - DNS는 상황에 따라 TCP/UDP 모두 가능하다. 그러나 UDP를 주로 사용한다.

- 중요한 3계층 프로토콜: **IPv4**. 4계층에서는 TCP/UDP 헤더를 고르지만, 3계층에서는 둘 다 IP헤더를 부착한다(**필수 헤더**)

- ARP, RARP도 3계층에 속한다. IP를 보고 MAC 주소를 찾기 때문
- 또 다른 3계층 프로토콜: ICMP(인터넷 제어 메시지 프로토콜), IGMP(멀티캐스트 관리 프로토콜)

### 프로토콜
- http: 웹 클라이언트와 웹 서버 사이에서 웹 문서를 평문으로 송수신. apache만으로 구축 가능

- https: 보안을 강화한 버전이 https(http + ssl/tls). 웹 문서를 암호화시켜서 송수신. apache + 인증서 파일로 구축 가능
    * SSL: 인증서 기반의 암호화

- shttp: WC와 WS 사이에서 암호화된 웹문서 송수신(http + ssh)
    * SSH: 키 기반의 암호화. 제한적으로 심플하게 암호화

- ftp: 평문의 파일 전송(포트번호 2개 사용)

- ftps: 전송되는 파일을 암호화. (ftp + ssl)

- sftp: 전송되는 파일을 암호화 (ftp + ssh)

- telnet: 원격 접속(평문)

- ssh: telnet + ssh

## wireshark
- 패킷 분석 + 패킷 수집

- sniffing --> 패킷 도청
- sniffer --> 패킷 스니핑 툴

- 무차별모드(Promiscuous mode)
    - 데이터링크 계층과 네트워크 계층의 **주소 필터링을 해제한 모드**
        - 2,3계층을 확인하지 않는다.
    - **주소 지정에 관계없이** 호스트의 프로세서에 모든 패킷을 전달
    - LAN카드에 존재(일반 모드, 무차별 모드)
    - 패킷이 CPU에 전달되면 분석을 위해 패킷 스니핑 애플리케이션에 넘겨짐
    - 유닉스 또는 리눅스에서는 ‘ifconfig eth0(LAN카드 이름) promisc’로 설정
    - Window에서는 winpcap(최신 버전: npcap) 설치로 무차별 모드 설정
        - wireshark 설치 시 같이 설치 가능

- 패킷을 감지해도 수신지 주소가 맞지 않아 일반적으로는 폐기한다. 패킷을 보기 위해선 무차별모드가 필요

- 온라인 실시간 강의의 경우 UDP 패킷이 많음
