## 다계층 스위치
- L2 스위치와 L3, L4, L7계층 스위치가 존재한다. 이 때, L2 스위치를 제외한 나머지는 다계층 스위치(Multilayer Switch)라고 한다.

- 사내망과 외부망을 구분해주는 장비로는 라우터를 사용한다!
    - 라우터는 경계에 위치하여 내부와 외부를 분리한다.

    ## 포워딩 포트 == 수신포트 결정

    ### L2 스위치
    - 수신지 MAC 주소를 기반으로 포워딩 포트(출력 포트)를 정한다.
    - 기준이 수신지의 MAC 주소

    ### L3 스위치
    - 수신지 IP를 보고 포워딩 포트를 결정할 수 있다.
    - 기준이 수신지의 IP 주소

    ### L4 스위치
    - L4 스위치는 수신지 IP와 수신지 포트 번호를 보고 출력 포트를 결정한다.
    - 기준이 수신지 IP + 수신지 Port

    ### L7 스위치
    - 7계층은 텍스트까지 확인.
    - 기준이 수신지 IP + 수신지 Port + content(Text)

    ## 기능으로 구분하는 스위치 장비

    ### L2 스위치
    - Forwarding: 포워딩(전달)

    ### L3 스위치
    - L2 스위치의 기능 + 라우팅    

    ### L4 스위치
    - L3 스위치 + 로드 밸런싱: 트래픽 균일 부하 분산
    - 주기적으로 서버들에게 ping, arp, http 패킷 등을 보내 상태 확인.
    - Health Check를 통해 서버가 다운되면 다른 서버로 트래픽을 보내주는 기능. --> **고가용성 보장. 서버 이중화 구성**
        - 전제조건: 동일한 기능을 하는 서버가 다수 존재해야 한다.
        - 주요 용도: FLB(방화벽 로드 밸런싱), SLB(서버 로드 밸런싱)
    - Active-Standby 구성
        - 일반적인 이중화 구성
        - Active를 주기적으로 확인
        - Active가 다운되면 St를 Active로 활성화시킬 수 있다.

    - Active-Active 구성
        - 두 장비가 동시에 트래픽 처리
        - 부하 분산: 두 장비가 역할을 나눠서 처리
        - 복잡한 구성: 세션 동기화, 장애 감지, VIP 공유 등 고려 요소가 많음
        - 성능 우선 환경에서 사용
    - 데이터 센터쪽에서 많이 사용
    - **서버가 여러 대이지만 기능이 서로 다르면 L3 스위치를 사용한다!**

    ### L7 스위치
    - L4 스위치 + Security + Content(7계층 Text) 인식
    - work station 접속자와 모바일 접속자의 환경을 다르게 설정할 수 있다.
        - 7계층 데이터를 확인할 수 있어 모바일 접속 주소와 워크 스테이션 접속 주소를 다르게 줄 수 있다.
    - Load Sharing: 트래픽 비균일 부하 분산
    - Security: 악성 코드들이 Content 안에 들어있어 이를 기반으로 접속 제한 등이 가능하다.
    - L7 스위치는 많이 사용되지 않는다.+



- 멀티레이터 스위치에서 라우팅은 **선택적 기능**

## 사내망 구성
- Access Layer
- Distribution Layer
- Core Layer

### Access Layer
- end point들의 집합점
- Access layer 스위치에 end point들이 모여서 하나의 망을 구성한다.(1.0, 2.0망 등)

- Access layer 장비로는 허브 또는 L2 스위치를 사용한다.

### Distribution Layer
- 망과 망 사이를 연결해주는 라우팅 기능을 탑재한 멀티 레이어 스위치를 사용.
- Access Layer들의 집합점

- 게이트웨이 역할(라우팅), 보안정책, NAT, MLSW(L3SW)

- 이중화 구성/회선 이중화: 같은 장비를 두 개 이상 설치하여 한 대가 고장나더라도 다른 장비로 경유할 수 있도록 하는 것. 안전성/가용성을 높임

- 멀티 레이어 대신 라우터가 올 수 있는가?
    - 올 수는 있지만, 라우터는 스위치보다 속도가 느리다. --> 전체적인 대역이 낮아진다.
    - 그러나, 그 반대는 불가능하다. (라우터 자리에 멀티레이어 스위치 불가능)
    - 라우터는 동기종망(LAN & LAN) 또는 이기종망(LAN & WAN) 연결 시 사용한다.
    - 스위치는 동기종망 연결 시 사용

- 서버와 연결된 스위치들은 정책 설정이 가능한 L3 이상의 멀티레이서 스위치를 사용한다.

### Core Layer
- 코어 층의 스위치는 라우팅 기능을 포함하지 않는다.
- **빌딩과 빌딩 사이**의 데이터를 주고 받을 수 있게 해주는 층
- 코어를 두지 않으면 빌딩 간에 데이터를 주고받기 위해서는 풀매쉬로 연결해야 하기 때문에 비용이 매우 비싸다. 또한, 확장이 용이하지 않고 관리가 어려워 코어가 필요하다!

    - 풀매쉬의 장단점
        - 속도 빠름
        - 장애 발생 시 우회 가능
        - 관리 어려움
        - 비용 높음
        - 확장 용이성 낮음

    - 코어의 장단점
        - 확장 용이
        - 비용 감소
        - 속도 풀매쉬보다 느림 --> 고속 전달 기능에 치중(기능 최소화). L2 장비와 같은 기능이지만 처리율에 차이가 있다.


### 소규모 사내망
- IPS에서 주는 공유기와 모뎀을 사용.
- 공유기가 게이트웨이 역할을 한다.
- 모뎀은 WAN 연결로 외부망을 취급

### 일반적인 가정집
- 모뎀 --> 공유기 --> PC에 직접 연결
- 공유기는 서버급 기능을 보유(3계층 기능)
    - 라우팅 --> 포워딩 기능 포함
    - DHCP: 유동 IP 환경. --> DHCP서버. 공유기의 DHCP 서버가 ip 할당
    - 사설IP를 제외한 IP를 공인 IP 주소라고 한다.
        - 기본적인 사설 IP
            1. A Class : 10.x.x.x
            2. B Class : 172.16.x.x ~ 172.31.x.x
            3. C Class : 192.168.x.x
        - 사설 IP는 인터넷에 접속할 수 없다!
        - 이를 가능하게 해주는 것이 NAT!
    - NAT



### NAT
- 회사에서의 NAT
- 공유기를 통과하는 순간 사설 --> 공인: SNAT
- 도착하는 서버쪽 서버 --> 사설: DNAT
- 네트워크 주소 변환
    - A ---> B 
        - 1.1.1.1 ---> 2.2.2.2, 공인 --> 공인
        - 3.3.3.3 ---> 10.1.1.10, 공인 --> 사설
        - 172.16..1.1 ---> 192.168.1.1, 사설 --> 사설
        - 192.168.1.10 ---> 5.5.5.5, 사설 --> 공인
    - 송신지 또는 수신지 IP 주소를 다른 주소로 변환
    - **Source NAT(SNAT)와 Destination NAT(DNAT)**
        - SNAT, DNAT 어떤 것이 진행되었는지가 중요

        - 인터넷 망의 라우팅 테이블은 공인 IP만 존재한다.
        - 데이터를 주고 받으려면 자신의 IP도 공인이어야 함
        - 공유기에 존재하는 NAT Table --> 임대 IP

        - 월 인터넷 사용료: 장비 임대료(공유기, 모뎀)
        - **사설 IP를 사용하면 비용 절감 + 보안이 가능하다!**

### PAT
- 포트 번호 변환
- 장점
    - IP 절약: 공인 IP 하나로 수천 개의 연결 처리 가능
    - 보안 강화: 내부 IP가 외부에 노출되지 않음
    - 유연한 네트워크 구성 ISP 변경 시에도 내부망은 그대로 유지 가능
    - 주소 뿐만 아니라 TCP/UDP의 포트 번호도 교환

- 단점
    - 특정 서비스 제약: VoIP, P2P 등 실시간 통신에 불리할 수 있음
    - 포트 충돌 가능성: 너무 많은 연결 시 포트 번호 중복 위험
    - 추적 어려움: 외부에서 내부 장비 식별이 어려움

### 방화벽

- IP주소와 Port 번호를 기반으로 방화벽 rule set(필터링 정책)에 따라 패킷 필터링을 수행하는 보안 장비
    - 접근 제어(Access Control)/패킷 필터링 (메인 기능)
        - 방화벽 관리자: 접근 제어 Rule을 생성하는 사람

    - NAT (Network Address Translation). 방화벽에 설정하는 경우가 많음
    - 액세스 기록 기능
    - 사용자 인증 (Authentication) 
    - 암호화 + 터널링 --> VPN 기능(별도의 라이센스 구매 필요)


- 기본적으로 방화벽을 기준으로 망을 3가지로 구분할 수 있다.
    1. 인터넷 망(Untrust Zone): 통신사 등 인터넷을 통해 외부에서 직접 접근이 가능한 구간.

    2. 인트라넷(Pricate/Trust Zone)
        -  물리적으로 분리된 망.
        - 접근 통제 시스템에 의해 인터넷으로 직접 접근이 불가능하도록 통제/차단되어 있는 구간

    3. DMZ(Public/Service Zone)
        - 외부망과 내부망 사이에 위치한 중간지점
        - 접근 통제 시스템을 통해 접근 통제 수행
        - 외부에 서비스를 제공하면서 내부 자원/시스템을 보호하기 위해 외내부망 사이에 접근 통제를 수행

    + 대외망(Extranet): 회사 대 회사로 서비스 연동이 필요한 경우 인터넷, 전용선, VPN 등으로 연동

- 방화벽은 송신지 IP --> 수신지 IP --> 수신지 포트 번호를 확인
- 방화벽은 3계층 장비



### 방화벽의 한계
- 내부망에서의 공격에 취약
    - 내부 네트워크에 존재하는 악의적인 공격을 막을 수 없음
    - 방화벽을 경유하지 않는 공격을 막을 수 없음 --> 라우팅 정책이 잘못 된 경우.
        - 내부 직원이 USB나 하드 디스크 등을 가져가는 행위는 막을 수 없다.
        - 기업 보안 장비를 별도로 사용해야 함
    - 방화벽에 방어 규칙에 포함되지 않는 공격을 막을 수 없음
        - **접근 제어에 대한 Role이 아닌 공격은 막을 수 없다.**
        - 제외된 규칙에 대해선 막지 못한다.

- 7계층 공격에 취약
    - 데이터에 실려 있는 악성 코드나 바이러스를 막을 수 없음
    - 메일에 첨부된 악성 코드를 막을 수 없음
- DoS와 DDoS 공격을 막을 수 없음

- 열린 포트를 통한 공격을 막을 수 없음

- 예전 방화벽 기준으로 공격 100개 중 40~50개 정도를 방어하는 정도로 방어율이 좋지 못하다.


### 침입 탐지 시스템 (IDS, Intrusion Detection System)
- 방화벽의 한계를 보완하기 위해 개발된 장비
- 내부망에 있는 비정상적인 트래픽을 **탐지**하는 시스템
    - 예시: 강의실에서 정상적인 수강생들은 질문을 하거나 강의를 열심히 들음
    <---> 비정상적인 수강생은 수업 중 계속 잠을 자거나 딴짓을 함
    - IDS는 이 트래픽이 활동하지 못하도록 **공격 패킷 차단을 할 수 있지만 하지 않고 탐지에 집중한다.**

- 공격을 탐지하고 관리자에게 공격 알림을 통해 공격에 대처할 수 있게 해 주는 보안 시스템
- 전달하는 패킷의 내용이나 로그를 분석하여 공격 여부를 탐지
    - **악성코드 탐지 가능**

- HIDS(Host-based IDS)와 NIDS(Network-based IDS)로 분류
    - HIDS는 특정 서버에 설치하고, NIDS는 이 서버에 공격자가 들어오는지를 찾는 것
    - 무결성 점검
    - 일반적으로 말하는 IDS는 NIDS. 스니핑 구성이 되어 있어야 함
        - 패킷이 들어올 때 IDS에 이 패킷들이 복사되어 있어야 탐지 가능
        - 공격으로 지정해둔 형태의 패킷이 들어왔을 때 IDS가 탐지
        - 탐지 후 관리자에게 알리면 관리자가 이 패킷을 차단하는 방식.
