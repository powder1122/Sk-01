## 데이터 중심 경향
- 데이터의 중심 경향은 데이터를 대표하는 값을 나타내는 척도
- 평균, 중앙값, 최빈값은 각각 데이터를 요약하는 방법으로 사용
- 각 방법은 서로 다른 상황에서 유용

### 평균
- 평균은 데이터를 모두 더한 뒤 데이터의 개수로 나눈 값으로 **데이터를 고르게 나눈 "중심값"**으로 볼 수 있다.

- 계산 방법: 평균= (데이터의 총합)/(데이터의 개수)

- 특징: 모든 데이터를 고려하므로 **데이터가 균일할 때 유용**
    - **극단값(이상치)가 포함되면 영향을 크게 받는다.**
        - 이 경우 평균을 중앙값으로 보기 어렵다.



### 중앙값(Median)
- 중앙값은 데이터를 크기 순으로 정렬했을 때 **가운데 위치한 값**으로, **데이터가 극단값의 영향을 받을 때 평균보다 대표성을 가짐**

- 계산 방법:
    1. 데이터를 크기 순으로 정렬
    2. 데이터 개수가 홀수면 가운데 값이 중앙값
    3. 데이터 개수가 짝우면 가운데 두 값을 평균낸 것이 중앙값

- 특징: **극단값의 영향을 받지 않음**

### 최빈값(Mode)
- 최빈값은 데이터에서 가장 자주 나타나는 값으로 **범주형 데이터나 특정 값의 빈도가 중요한 경우 유용**

- 계산 방법: 
    1. 데이터에서 각 값의 빈도를 계산
    2. 가장 빈도가 높은 값을 선택

- 특징: 하나 이상의 최빈값이 있을 수 있음. 연속형 데이터에서는 잘 사용되지 않음.(점수가 모두 다르면 최빈값이 없다.)

### 비교와 활용

- 평균
    - 사용 목적: 데이터를 고르게 대표
    - 유용한 상황: 데이터가 균일할 때(정규분포)

- 중앙값
    - 사용 목적: 극단값의 영향을 배제
    - 유용한 상황: 이상치가 포함된 경우

- 최빈값
    - 사용 목적: 가장 흔한 값을 찾음
    - 유용한 상황: 범주형 데이터 분석 시

## 데이터 분산성
- 데이터 분산성은 데이터가 **평균을 기준으로 얼마나 퍼져있는지***를 나타낸다.
- 분산, 표준편차, 사분위수는 각각 다른 방식으로 데이터를 요약하여 데이터의 분포와 특성을 파악하는 데 유용

    - 각 데이터가 평균에서 얼마나 떨어져 있는지를 나타내는 값
    - 편차들을 하나로 묶어 대표값처럼 나타낸 것
    - 편차들을 제곱해서 평균내고 루트를 씌운 값(제곱근)
    

### 분산
- 분산(Variance): 분산은 데이터가 평균을 기준으로 얼마나 퍼져 있는지를 측정하는 값 --> **값이 클수록 데이터가 더 퍼져 있음을 의미**

- 계산 방법
    1. 각 데이터 값에서 평균을 뺀다(**편차**)
    2. 편차를 제곱한다.(음수를 없애기 위해)
    3. 제곱된 편차의 **평균**을 계산한다.



### 표준편차
- 분산의 제곱근. *표준편차 = sqrt(분산)*
- 데이터의 퍼짐 정도를 원래 데이터와 동일한 단위로 나타냄
- 특징
    - 분산과 달리 데이터와 동일한 단위를 사용하므로 해석하기 쉬움
    - 값이 작으면 데이터가 평균에 가까이 모여 있고, 값이 크면 데이터가 퍼져 있음을 의미

### 사분위수
- 사분위수는 데이터를 네 개의 구간으로 나누는 값으로, 데이터의 분포를 요약하는 데 사용
    - Q1 (1사분위수): 하위 25% 데이터 경계값.
    - Q2 (2사분위수, 중앙값): 하위 50% 데이터 경계값.
    - Q3 (3사분위수): 하위 75% 데이터 경계값.
    - IQR (사분범위): 𝑄3−𝑄1로, 중간 50% 데이터의 범위를 나타냄

- 특징
    - 분산과 달리 데이터와 동일한 단위를 사용하므로 해석하기 쉬움
    - Q3 - Q1 **값이 작으면 데이터가 평균에 가까이 모여 있고, 값이 크면 데이터가 퍼져 있음을 의미**

- 계산 방법
    - 데이터를 오름차순으로 정렬
    - Q1, Q2, Q3를 계산

- 특징: 이상치를 파악하는 데 유용(이상치는 Q1 - 1.5*IQR보다 작거나 Q3 + 1.5*𝐼𝑄𝑅보다 큰 값으로 정의)

### 비교와 활용

- 분산
    - 정의 및 특징: 데이터의 퍼짐 정도를 제곱 단위로 측정
    - 활용 상황: 데이터를 수학적으로 분석할 때

- 표준편차
    - 정의 및 특징: 데이터의 퍼짐 정도를 원래 단위로 측정
    - 활용 상황: 평균과 함께 데이터의 분포를 해석할 때

- 사분위수
    - 정의 및 특징: 데이터 분포의 경계값 및 중간 50% 범위를 측정
    - 활용 상황 데이터 분포의 범위를 시각화 하거나 이상치를 파악할 때

## 데이터 분포

- 데이터 분포는 데이터가 **공간적으로 어떻게 퍼져 있는지**를 나타냄
- **정규분포, 왜도, 첨도**는 데이터의 분포 특성을 이해하는 데 중요한 개념


### 정규분포
- 정규분포는 데이터가 평균을 중심으로 대칭적으로 분포된 종 모양의 분포. **종 모양**
- 데이터의 대부분이 평균 근처에 몰려 있고, 평균에서 멀어질수록 빈도가 줄어듦
- 특징
    - 평균, 중앙값, 최빈값이 **동일**
    - 데이터의 68%는 평균 +-1 표준편차 내에 있음
    - 데이터의 95%는 평균 +-2 표준편차 내에 있음
    

### 왜도
- 왜도는 데이터의 **비대칭성을 측정하는 값**
- 분포가 한쪽으로 치우쳐 있으면 왜도가 발생
    - 왜도 > 0 (양의 왜도): 데이터가 오른쪽으로 꼬리가 길게 늘어남 --> 평균이 높고, 중앙 값을 기준으로 최빈값이 낮은 모양이다.
    - 왜도 < 0 (음의 왜도): 데이터가 왼쪽으로 꼬리가 길게 늘어남. --> 중앙값을 기준으로 평균이 낮고, 최빈값이 높은 모양

- 특징
    - 왜도를 통해 분포의 비대칭성을 파악.
    - 양의 왜도: 평균 > 중앙값 > 최빈값.
    - 음의 왜도: 평균 < 중앙값 < 최빈값.

### 첨도
- 첨도는 분포의 꼭대기가 얼마나 뾰족하거나 평평한지를 측정하는 값
    - 첨도 > 3 (Leptokurtic): 꼭대기가 뾰족하고 꼬리가 두꺼움. (극단값이 자주 발생)
    - 첨도 = 3 (Mesokurtic): 정규분포와 비슷한 모양.
    - 첨도 < 3 (Platykurtic): 꼭대기가 평평하고 꼬리가 얇음. (극단값이 적음)

- 첨도가 크면 극단값(꼬리)이 많아 분석에 주의가 필요.
- 첨도가 작으면 데이터가 평균 근처에 고르게 분포.

### 비교와 활용

- 정규분포
    - 정의 및 특징: 평균을 중심으로 대칭적 분포
    - 활용 상황: 자연현상 분석, 통계 분석

- 왜도
    - 정의 및 특징: 분포의 비대칭성 측정
    - 활용 상황: 데이터 치우침 확인 (예: 주택 가격 분석)

- 첨도
    - 정의 및 특징: 분포의 꼭대기 뾰족함과 꼬리 두께 측정
    - 활용 상황: 극단값의 빈도 및 위험 분석 (예: 주식 수익률)

## 데이터 통합 및 조인
- 데이터 분석을 할 때, 여러 데이터셋을 하나로 합쳐야 할 때가 많음
- 이 과정에서 merge와 concat은 매우 중요한 역할을 한다.
- 두 방법 모두 **데이터를 합치는 기능**을 하지만, 어떤 기준으로 데이터를 합치는 지에 따라 사용 방식이 달라진다.

### Merge
- merge
    - 기준 열을 기반으로 데이터를 합치기(옆으로 붙이기)
    - 두 데이터프레임의 **공통 열(Key)를 기준으로 데이터를 합칠 수 있다.** --> 데이터를 안전하게 붙일 수 있다. (문제 발생 여지가 적어진다.)
    - SQL의 JOIN과 유사
        - Inner Join (교집합): 두 데이터프레임의 공통된 값만 유지
        - Outer Join (합집합): 공통된 값이 없어도 데이터 유지 (NaN으로 채움).
        - Left Join: 왼쪽 데이터프레임의 값을 모두 유지.
        - Right Join: 오른쪽 데이터프레임의 값을 모두 유지.



### Concat
- 데이터 방향으로 쌓기
    - 데이터를 방향(축)으로 단순히 이어 붙임(행 방향: axis=0, 열 방향:axis=1)
    - 기준 열 없이 데이터를 추가하고 싶을 때 유용

## 데이터 분포 확인


### 히스토그램
- 히스토그램은 데이터의 분포를 나타내는 데 사용
- 연속형 데이터를 구간(bin)으로 나누고 각 구간에 해당하는 데이터의 빈도를 보여줌
- 사용 목적
    - 데이터 값이 어떻게 퍼져 있는지 확인(분포 확인)
    - 데이터의 대략적인 형태(정규분포, 치우침 등)파악
    - 이상치(outlier) 탐지

### 박스플롯
- 데이터의 중앙값, 사분위수, 이상치를 시각화, 데이터 분포를 요약하고, 이상치를 쉽게 탐지할 수 있는 도구

- 사용 목적: 데이터의 중앙값과 변동성 확인, 이상치 탐지, 데이터 간 비교

- 박스플롯의 주요 구성 요소
    - 중앙값: 데이터의 중간값(박스의 중앙선)
    - 1사분위수: 하위 25%데이터 값
    - 3사분위수: 상위 25%데이터 값
    - IQR: Q3 - Q1(데이터 변동성 측정)
    - 이상치: Q1 - 1.5×IQR 이하, 또는 Q3 + 1.5×IQR 이상 값

## 데이터 탐색적 분석(EDA)을 위한 시각화 기법
- EDA(탐색적 데이터 분석)는 데이터를 시각적으로 분석하여 **패턴, 관계, 이상치, 분포** 등을 탐구하는 과정
    - 관계 속에서 특정 패턴이 있다는 가설을 세워 확인하는 과정이 EDA이다.
    - 이상치를 보고 이 값을 삭제할지, 변환할지, 포함할지 등을 생각한다.
    - 분포를 보고 어느쪽으로 치우쳐져 있는지, 이를 바탕으로 예측을 할 수 있는지 확인
    - *찾기 쉽지 않다.*

- 다양한 시각화 기법을 통해 데이터를 더 깊이 이해하고 분석 방향을 설정할 수 있음

### 선점도
- 인풋과 아웃풋 처리
- 두 데이터 간의 관계 확인 목적
- sns.scatterplot(x='',y='',data=data)

### 히트맵
- 변수 간 상관관계 분석!
- 여러 변수 간의 상관관계를 확인하는 것이 목적


### 페어플롯
- 변수 간 관계를 선점도로!
- 변수 간 관계를 한눈에 확인
- 데이터셋의 모든 수치형 변수 간의 관계를 시각화


### 카운터플롯
- 빈도 수 확인!
- **범주형 데이터 시각화**

### 시각화 기법의 선택 기준

- 산점도 (Scatter Plot) 
    - 사용 목적: 두 변수 간의 관계를 시각적으로 탐구
- 히트맵 (Heatmap) 
    - 사용 목적: 여러 변수 간의 상관관계를 시각화
- 박스플롯 (Boxplot) 
    - 사용 목적: 데이터의 분포와 이상치 탐지
- 페어플롯 (Pair Plot)
    - 사용 목적: 데이터셋의 모든 변수 간 관계 탐색
- 카운트플롯 (Count Plot)
    - 사용 목적: 범주형 데이터의 빈도 탐색

## EDA
- 데이터 분석 과정에서 데이터를 이해하고 통찰을 얻기 위한 초기 분석 단계
    - 데이터셋의 구조와 특성을 탐색
    - 문제가 되는 부분을 확인
    - **분석 방향(관계)을 설정하는 데 필수적인 과정**

- 데이터를 알아가는 과정

- EDA의 주요 목적

    - **!!데이터의 기본 특성 이해!!**
        - 데이터를 시각적(그래프)으로 관찰하고, 수치로 요약하고 데이터 분포, 이상치, 결측치 등을 확인(중심값 --> 평균,중앙,최빈. + 왜도,첨도)

    - 데이터 정리
        - **결측값이나 이상치를 처리**하고 데이터를 **분석 가능한** 상태로 준비

    - 가설 수립 및 데이터 기반 의사 결정
        - 데이터를 관찰하며 분석 방향과 **가설을 수립**. 특성과 특성관의 규칙, 패턴을 찾는다!

    - 문제 발견
        - 분석 중 문제가 될 수 있는 요소를 확인(이상치, 널값)

### EDA의 주요 단계

1. 데이터의 기본 특성 이해
    - 데이터 불러오기 및 기본 정보 확인데이터셋의 크기(행과 열의 수), 변수명, 데이터 유형(숫자, 문자 등)을 확인

    - Python에서 pandas 라이브러리의 head(), info(), describe() 등을 사용

2. 결측 값과 이상치 확인
    - 결측 값: 데이터가 비어 있는 경우 (NaN 값. isnull())
    - 이상치: 비정상적으로 높거나 낮은 값.
    - 이를 확인하고 처리 방법을 결정: 삭제, 대체

3. 데이터 분포 확인
    - 변수의 분포를 이해하기 위해 히스토그램이나 커널 밀도 그래프를 사용
    - 데이터가 정규 분포(normal distribution)를 따르는 지 확인할 수 있음

4. 변수 간 관계 탐색
    - 두 변수 사이의 **상관관계**를 확인하거나, 그룹별 통계량을 계산
    - 데이터를 의미있는 데이터로 만들어주기 위함!!

5. 시각화
    - 시각화는 데이터를 더 직관적으로 이해할 수 있게 도와줌
    - 사용 도구: Matplotlib, Seaborn



### EDA의 결과물
- EDA를 통해 데이터를 충분히 이해하면 다음을 얻을 수 있다.
    - 데이터셋의 구조와 **특성에 대한 이해**(중요!!)
    - **이상치와 결측치 처리 계획** (무조건 지우는 게 답이 아니다!)
    - 데이터 분포와 변수 간 **관계**에 대한 통찰
    - 분석 방향성 설정 및 **가설** 정립(핵심!!)

## 데이터의 기본 구조 파악
- 데이터 분석을 시작하기 전에 데이터의 기본 구조를 이해하는 것은 매우 중요하다.
- 데이터를 구조적으로 이해하면 분석 목표에 맞는 전략을 세우고, 데이터를 올바르게 처리할 수 있다.
- 데이터의 기본 구조를 파악하는 과정
    - 데이터의 모양(형태)
    - 데이터 타입
    - 변수의 분포 및 관계
    - 결측 값 유무

### 데이터의 기본 구조
1. 데이터셋의 크기
    - 데이터셋이 몇 개의 행과 열로 구성되어 있는지 확인
    - 데이터를 요약할 때 데이터의 전반적인 크기를 알고 있어야 효율적으로 작업할 수 있음

2. 데이터 타입
    - 각 열의 데이터 유형을 이해 --> 데이터 유형은 분석 방법에 영향을 미침
        - 수치형 데이터(Numeric): 정수형, 부동소수점
        - 범주형 데이터(Categorical): 특정 그룹이나 클래스를 나타냄
        - 문자형 데이터(String): 텍스트 데이터
        - 날짜/시간 데이터(DateTime): 시간 기반 데이터

3. 변수의 이름과 의미: 데이터셋에 포함된 변수(컬럼)의 이름과 의미 파악

4. 결측값: 데이터셋에 비어 있는 값이 있는지 확인 --> 분석 결과에 영향을 미칠 수 있으므로 반드시 처리

5. 데이터 분포: 변수의 분포를 파악하여 데이터가 어떻게 구성되어 있는지 이해

6. 중복값: 동일한 행이 여러 번 반복되는지 확인


### 데이터의 기본 구조를 파악한 후 할 수 있는 일
- 분석 전략 설정
    - 각 변수의 타입에 따라 분석 방법을 설정
    - 예: 수치형 데이터 → 평균, 중앙값 계산.
    - 범주형 데이터 → **빈도 분석**.

- 데이터 전처리
    - 결측값 처리: 평균값으로 대체, 삭제, 또는 **예측값**으로 대체.(예측 모델을 만들어 값을 예측한다.)
    - 데이터 **타입 변환**: 필요에 따라 숫자를 범주형으로 변환하거나 반대 작업 수행.

- 데이터 시각화
    - 변수 간 **관계**를 시각적으로 탐색
    - 예: 산점도(scatter plot)로 수치형 데이터 관계 확인.
    - 막대그래프(bar chart)로 범주형 데이터 분포 확인
    - 비지도 학습에서 데이터 분포를 확인하여 값을 처리할 수 있다.

## 주요 기초 통계량


### 1. 중심 경향성
- 데이터의 대표 값을 나타냄
    - 평균: 데이터의 합을 데이터 개수로 나눈 값. 대부분의 데이터가 이 근처에 분포한다고 볼 수 있음
    - 중앙값: 데이터를 크기 순으로 정렬했을 때 가운데 위치한 값. 데이터가 치우쳐 있을 때 평균 대신 중앙값을 사용
    - 최빈값: 데이터에서 가장 자주 나타나는 값. 특정 값이 얼마나 자주 나타나는지 확인
    

### 2. 산포도
- 데이터가 얼마나 흩어져 있는지 측정
    - 분산: 데이터가 평균으로부터 얼마나 떨어져있는지 나타내는 값
    - 표준편차: 분산의 제곱근으로, 데이터의 흩어짐 정도를 나타냄. 표준편차가 크면 데이터의 변동성이 크다는 뜻
    - 범위: 데이터의 최대값과 최소값의 차이



### 3. 데이터 분포 특성
- 데이터 분포의 모양과 특성을 이해

    ## 왜도
    - 데이터 분포가 비대칭인지 나타냄. 
        - 양수: 오른쪽으로 꼬리가 긴 분포. 평균이 중앙값보다 높고, 최빈값이 중앙값보다 낮다

        - 음수: 왼쪽으로 꼬리가 긴 분포. 평균이 중앙값보다 낮고, 최빈값이 중앙값보다 크다.

    ## 첨도
    - 분포의 뾰족한 정도를 나타냄
        - 양수: 분포가 뾰족함
        - 음수: 분포가 평평함
