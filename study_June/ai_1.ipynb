```python
from bs4 import BeautifulSoup
```

```python
html_doc = """
<html>
    <head>
        <title>Example Page</title>
    </head>
<body>
    <h1>Welcome to Web Scraping</h1>
    <p class="description">This is an example paragraph.</p>
    <a href="https://example.com1">Visit Example1</a>
    <a href="https://example.com2">Visit Example2</a>
    <a href="https://example.com3">Visit Example3</a>
    <a href="https://example.com4">Visit Example4</a>
</body>
</html>
"""
```

```python
#BeautifulSoup 객체를 생성
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
#string으로 텍스트만 출력
print(soup.title.string)
print(soup.h1.string)  
print(soup.a['href'])   #대괄호를 통해 속성 접근.
print(soup.p)
print(soup.p['class'])  #클래스 속성의 경우 리스트를 반환
```

```python
#find(), find_all() 메서드
#find(tag선택): 처음 만난 것 하나 반환
#find_all(tag선택): 여러 개 리스트 형식으로 반환
a_tag = soup.find('a')
print(a_tag)
a_tags = soup.find_all('a')
print(a_tags)
```

```python
links = []
for a_tag in a_tags:
    print(a_tag['href'])
    links.append(a_tag['href'])
    
print(links)
```

```python
# 선택자. 클래스 안에서 사용할 수 있돌고 할 때는 '.' 추가. class 의미. 더 간단하게 사용 가능
description = soup.select_one('.description') 
print(description.string)
```

```python
#requests 사용하여 문서 가져오기 
url = "https://news.ycombinator.com/"
```

```python
response = requests.get(url)
```

```python
print(response.text[:500])
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.find_all('a'))
```

```python
for item in soup.find_all('a'):
    href = item.get('href')
    #데이터를 가져오지 못하는 경우가 있다.
    if href:
        print(item.get_text(strip=True))
        print(href)
```

```python
#크롤링 막아둔 네이버 뉴스 페이지 이용해보기.
#헤더값을 이용.
url = "https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=105"

#요청 헤더 설정(크롤링 방지 우회를 위해 User-Agent 추가)
#로봇이 아니다~
headers = {
    "User-Agent": "Mozilla/5.0(Windows NT 10.0;Win64;x64) AppleWebkit/537.36(KHTML, like Gacko)Chrome/114/0/0/0 Safari/537.36"
}

response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')#원본 데이터
```

```python
news_items = soup.select("ul.type06_headline li dt:not(.photo) a")  #ul태그 밑에 li 안쪽의 a태그
```

```python
for item in news_items:
    title = item.get_text(strip=True)
    link = item['href']
    print('title: ', title)
    print('link: ', link)
    
```

```python
#특정 키를 입력(검색하기)
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys #키를 처리하는 함수 
import time #서버에서 응답대기를 위해 sleep() 이용 
```

```python
driver = webdriver.Chrome()

driver.get('https://www.google.com')

#검색창 찾기
search_box = driver.find_element(By.NAME, 'q') #NAME = 'q'
#검색어 처리 
search_box.send_keys('python') #python 검색어
#엔터
search_box.send_keys(Keys.RETURN) 

#결과가 나올 때 까지 대기
time.sleep(10) #5초 대기

results = driver.find_element(By.CSS_SELECTOR, "h3 > a")
results.click()
print(results)
```

```python
driver = webdriver.Chrome()

url = "https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=105"

driver.get(url)

#결과가 나올 때 까지 대기
time.sleep(3) #5초 대기

results = driver.find_elements(By.CSS_SELECTOR, 'ul.type06_headline li dt:not(.photo) a')
print(results)
for item in results:
    print(item.text)
    print(item.get_attribute('href'))

driver.quit()


```
