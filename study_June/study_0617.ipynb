## 데이터 전처리의 중요성
### 데이터 전처리란?
- **데이터를 분석**하거나 **머신러닝 모델**에 사용하기 전에 깨끗하고 유용한 상태로 준비하는 과정(Data --> 시각화)
- 데이터 수집 후 비어 있는 값(결측값, 누락된 값), 엉뚱한 값(오류, 이상치), 필요 없는 값(중복)이 있다면, 이를 수정하고 관리하는 과정
    - --> 데이터 전처리는 데이터를 '사용하기 좋게' 만드는 청소 과정

    ### 왜 필요한가?
    - 잘못된 데이터를 그대로 사용하면, 분석 결과나 모델이 엉뚱한 답을 내놓을 수 있기 때문
    - 데이터 전처리는 요리 전에 재료를 씻고 다듬는 것과 같다. 깨끗한 데이터로 분석해야 좋은 결과를 얻을 수 있다.
    - *학습하는 데이터의 표본이 굉장히 중요하다!*


- 데이터 전처리는 데이터의 품질을 높이고, 분석이나 모델링의 결과를 신뢰할 수 있게 해 줌.
- 데이터를 잘 정리하지 앟ㄴ으면 모델 성능에 큰 영향을 미치며(잘못된 결과 초래), 시간과 비용이 낭비될 수 있다.
- ** 전처리는 단순한 작업이 아니라 성공적인 데이터 프로젝트의 필수적인 단계

## 데이터 유형
- 정형 데이터(Structured Data)
- 비정형 데이터(Unstructured Data)
- 반정형 데이터(Semi-Structured Data)

### 정형 데이터
- 정형 데이터는 행과 열로 구성되어있는 잘 정리된 데이터 --> 데이터베이스나 엑셀 파일 
- 특징: 데이터가 표처럼 정렬되어 있다. 각 열에 이름(필드명)과 데이터 유형(숫자, 문자열 등)이 정해져 있음. **행단위로 처리**
- 예시: 은행의 고객 정보 데이터 --> 열: 이름, 나이, 계좌번호, 잔액. 행: 개별 고객의 정보.


### 비정형 데이터
- 고정된 구조가 없는, 정리되지 않은 데이터 --> **이미지, 동영상, 텍스트 파일**
- 특징: 데이터의 모양이나 구조가 일정하지 않고 분석하려면 먼저 데이터를 가공해야함
- 예시:
    - 텍스트 데이터: 이메일(제목, 내용), 채팅 기록, 문서 파일
    - 이미지 데이터: 사진, 스캔한 문서
    - 동영상 데이터: 유튜브 영상, 감시 카메라 영상
    - 오디오 데이터: 음성 녹음 파일 및 음악 파일 사용

### 반정형 데이터
-  반정형 데이터는 정형 데이터와 비정형 데이터의 중간 단계로 데이터에 구조는 있느지만, 정형 데이터처럼 엄격하지는 않다. --> 대표적으로 JSON과 XML 같은 파일
- 특징: 데이터에 구조가 있지만, 형태가 유연함.
- 사람이 읽을 수 있으며, 컴퓨터도 처리 가능
- 예시: json을 xml, 웹 로그 데이터

- 데이터 전처리는 데이터 전처리는 데이터를 분석이나 모델링에 사용하기 위해 깨끗하고 유용하게 준비하는 단계
- 데이터 수집 → 정제 → 변환 → 통합의 순서로 진행'
- 각 단계는 데이터의 *8품질과 활용도를 높이는 데 초점**이 맞춰져 있음

### 결측치 처리

- 결측치: 데이터셋에서 특정 값이 누락된 상태. **결측치는 데이터 분석 결과에 영향을 줄 수 있으므로 반드시 처리해야 함

- 결측치를 처리하는 이유:
    1. 분석 왜곡 방지: 결측값을 그대로 두면 평균, 합계 같은 계산이 왜곡될 수 있음
    2. 모델 학습 방해 방지: 머신러닝 모델은 걸측값을 포함한 데이터를 처리하지 못하는 경우가 많이 발생
    3. 데이터 품질 향상: 결측치를 처리하면 데이터가 더 신뢰할 수 있게 됨

- **결측치 제거**
    - 사용 상황
        - 결측치가 데이터셋에서 차지하는 비율이 매우 적을 때
        - 해당 열이나 행이 분석에 큰 영향을 주지 않을 때
    
    - 방법
        - 열 제거: 특정 열에 결측값이 너무 많다면, 해당 열 자체를 삭제
        - df.dropna(axis=1)
        - 행 제거: 특정 행에 결측값이 있으면, 해당 행을 삭제
        - df.dropna()

- **결측치 대체**
    - 사용 상황
        - 데이터의 양이 적거나, 결측치가 무작정 삭제되면 데이터 손실이 너무 클 때

    - 방법
        - 고정 값으로 대체. 특정 값(0,평균,중앙값 등)으로 결측치를 채움
        - 통계 기반 대체:
            1. **평균값**으로 대체: 수치형 데이터에서 결측값을 해당 열의 평균으로 대체
                - df[''].fillna(df[''].mean())
            2. **중간값**으로 대체: 평균이 극단값에 영향을 받는 경우, 중간값 사용
                - df[''].fillna(df[''].median())
            3. **최빈값**으로 대체: 범주형 데이터에서 가장 많이 등장한 값으로 결측치를 대체
                - df[''].fillna(df[''].mode()[0])
        - 예측 기반 대체: 머신러닝 모델을 사용해 결측치를 예측하여 채움


- 데이터 손실을 최소화하고 분석에 미치는 영향을 줄이려면
    - 결측치의 양, 데이터의 중요성, 그리고 분석 목표를 고려해야 한다.
    - 간단한 데이터셋에서는 제거와 평균/중간값 대체가 유용
    - 하지만, 복잡한 데이터에서는 머신러닝 기반 대체를 활용

- **원본 데이터의 사이즈에 영향을 많이 받는다!!**

### 이상치 처리

- 이상치: **데이터셋에서 다른 값들과 크게 동떨어져 있는 값**. 이상치는 분석 결과를 왜곡하거나 모델의 성능을 저하시킬 수 있으므로 적절히 처리해야 함.

- 이상치 처리의 필요성
    - 데이터 왜곡 방지: 이상치는 평균, 분산, 상관관계 등의 통계값에 큰 영향을 미칠 수 있음
    - 모델 성능 향상: 머신러닝 모델은 이상치로 인해 잘못된 패턴을 학습할 수 있음
    - 데이터 신뢰성 확보: 이상치를 처리하면 데이터의 품질이 높아짐

- 이상치 처리
    - IQR(Interquartile Range) 방법
        - IQR은 데이터의 중간 50%를 포함하는 범위를 이용해 이상치를 탐지하는 방법
        - 비대칭 분포 데이터에 적합
        - 극단값에 영향을 덜 받음
        - Q1(1사분위수): 히위 25%의 값
        - Q3(3사분위수): 상위 75의 값
        - IQR = Q3-Q1
        - 이상치 기준
            - 아래쪽: Q1 - 1.5*IQR보다 작은 값
            - 위쪽: Q3+ 1.5*IQR보다 큰 값

    - Z-Score 방법
        - Z-Score는 데이터의 값이 평균에서 얼마나 떨어져 있는지를 표준 편차 단위로 나타냄
            - 정규분포를 따르는 데이터에 적합
            - 계산이 간단함
        
        - Z= (X−μ)/σ
            - 𝑋: 개별 데이터 값
            - 𝜇: 평균.
            - 𝜎: 표준 편차.
        - 이상치 기준
            - 일반적으로 ∣𝑍∣>3이면 이상치로 간주
            


### 중복 데이터
- 중복 데이터는 데이터셋에서 동일한 행(row)이 반복적으로 나타나는 것

- 중복 데이터가 있으면 분석 결과가 왜곡될 수 있으며, 특히 데이터의 크기가 큰 경우 처리 속도와 메모리 사용량에 영향을 줄 수 있음

- 중복 데이터는 다음과 같이 나타날 수 있다.
    - 완전 중복: 모든 열의 값이 동일한 행

    - 부분 중복: 특정 열의 값만 동일한 행

- 중복의 기준: **행(row)**. DateFrame으로 다루기 때문에, 행을 기준으로 한다. (Table : 행 컨트롤)

- 중복이 나타날 경우, 특정 기준으로 하나만 남기고 나머지는 전부 삭제하는 등의 처리를 할 수 있다.

- 중복 확인하기
    - df.duplicated(): 중복이면 True

- 중복 삭제하기:
    - drop.duplicates(keep='first or last'): 중복 데이터 중 첫 번째 혹은 마지막 데이터 유지(행의 모든 열 값이 같은 경우)

    - drop.duplicates(subset=['열 이름'], keep=''): 특정 열 값이 같으면 삭제.

    - drop.duplicates(keep=False): 데이터를 유지하지 않고 모두 삭제.



## 데이터 변환 및 스케일링
- 데이터 변환(로그 변환, 스퀘어 루트 변환)
- 데이터 스케일링(표준화, 정규화)
- 범주형 데이터 처리(레이블 인코딩, 원-핫 인코딩)
    - 레이블 등의 데이터를 숫자형 데이터로 변환하는 작업
- 왜 필요한지에 집중하여 학습

### 데이터 변환
- 데이터의 분포를 변화시켜(눌러놓고) 분석이나 모델링에 유리한 형태로 만드는 과정
- 로그 변환과 스퀘어 루트 변환은 데이터를 처리할 때 자주 사용되는 기법
- 특히 데이터의 왜도(Skewness)를 줄이고 더 정규 분포에 가깝게 만들기 위해 활용

- 로그 변환:
    - 데이터 값의 로그(log)를 취하는 방법
    - 데이터 값이 매우 크거나, 몇 개의 값이 지나치게 크고 나머지는 작은 경우 --> (오른쪽으로 치우친 분포)를 다룰 때 효과적
        - 메출 데이터가 [1000, 2000, 5000, 10000, 100000]일 경우, 오른쪽에서 급격하게 올라가게 된다. 이런 경우 로그 변환을 하면 평탄한 형태로 그래프를 그릴 수 있다.

- 로그 변환을 하면
    - 큰 값은 축소되고, 작은 값은 상대적으로 더 커져 데이터 분포를 평탄하게 만들어 준다.

- 장점
    - 큰 값의 영향을 줄이고 작은 값의 상대적 차이를 부각할 수 있다.
    - 분포의 왜도를 줄이고, 모델링 시 더 나은 예측 성능을 얻을 수 있음

- 주의점
    - 데이터에 0이나 음수가 있으면 로그를 계산할 수 없기 때문에, **변환 전에 값에 작은 상수(예:1)를 더해야 한다.**

### 스퀘어 루트 변환
- 데이터의 제곱근을 구하는 방법
- 로그 변환보다는 덜 극단적인 방법
- 데이터의 크기를 축소하거나 오른쪽 치우침을 완화하는 데 사용

- 몇몇 값이 너무 커서 데이터가 불균형하게 보일 수 있는 문제를 해결

- 장점
    - 데이터의 스케일을 줄이면서도 값 간의 상대적 관계를 유지
    - 비선형적 관계를 일부 선형화 할 수 있음

- 주의점
    - 음수 값에 사용 불가(**음수에는 제곱근이 실수로 정의되지 않음**)
    

### 로그 변환과 스퀘어 루트 변환 비교

- 로그 변환
    - 적용 대상: 매우 큰 값, 오른쪽 치우친 분포
    - 효과: 큰 값을 더 크게 줄임
    - 수식: y=log(x)
    - 제약 조건: 음수/0에는 사용 불가

- 스퀘어 루트 변환
    - 적용 대상: 작은 값과 큰 값이 공존하는 분포
    - 효과: 큰 값을 약간 줄임
    - 수식: y = sqrt(x)
    - 제약 조건: 음수에는 사용 불가


- 정리
    - 로그 변환은 매우 큰 값을 다룰 때 효과적
    - 스퀘어 루트 변환은 큰 값과 작은 값의 균형을 맞추는 데 유용
    - 두 방법 모두 데이터의 분포를 조정하고, 모델 학습 성능을 개선하는 데 중요한 역할


### 데이터 스케일링: 표준화와 정규화
- 데이터 스케일링은 데이터의 값 범위를 조정하는 과정
- **머신러닝 모델이 데이터의 크기 차이에 영향을 덜 받게 하기 위해 중요함**
- 특히 변수의 단위나 크기가 다를 때, 스케일링은 모델 학습에 있어 필수적인 과정 


- 표준화(Standardization)
    - **표준화는 데이터의 값을 평균이 0, 표준편차가 1이 되도록 변환하는 방법**
        - *평균이 0,표준편차가 1이다 --> 전체의 68%가 -1~1사이의 값이다.*
        - 데이터를 똑같은 기준선에 맞춰서 공정하게 비교하게 해주는 과정
        - 평균: 데이터들의 중심값
        - 표준편차: 데이터가 평균에서 얼마나 떨어져있는지 측정한 값
    - **데이터의 분포 모양을 바꾸지는 않고, 단위 차이를 없애 주는 것**

    - 장점
        - 데이터가 서로 다른 범위와 단위를 가지더라도, 동일한 기준으로 변환
        - 특히, 거리 기반 알고리즘 (KNN< SVM)에서 효과적

    - 주의점
        - 데이터가 정규 분포를 따르지 않더라도 표준화를 적용할 수 있지만, 분포의 왜도는 남아있을 수 있다.

- 정규화
    - 정규화는 데이터를 0과 1사이로 변ㄴ환하는 방법
    - 값의 크기를 줄이고, 데이터를 일정한 스케일로 맞춤
    - 수식: Min-Max Scaling

    - 장점
        - 데이터가 일정 범위 내에 있어, 스케일에 민감한 알고리즘(신경망, 회귀)에서 적합
        - 이상치(outlier)의 영향을 줄이는 데도 유용함

    - 주의점
        - 데이터에 이상치가 있을 경우, 최대값과 최소값이 크게 왜곡될 수 있음 **--> 데이터 스케일 전 이상치에 대한 처리가 선행되어야 한다!**


### 표준화와 정규화 비교

- 표준화
    - 적용 후 범위: 평균 0, 표준편차 1
    - 사용 시기: 정규 분포에 가까운 데이터
    - 알고리즘 적용: 거리 기반 알고리즘(KNN, SVM 등)
    - 이상치에 대한 민감도: 덜 민감
    - 사용 사례: 병원 데이터 분석. 데이터가 모두 다른 단위를 가진다. **단위를 없앨 때-->표준화**

- 정규화
    - 적용 후 범위: 0에서 1사이
    - 사용 시기: 값의 범위가 클 때
    - 알고리즘 적용: 신경망, 회귀 등
    - 이상치에 대한 민감도: 더 민감
    - 사용 사례: 이미지 데이터 전처리. 이미지 픽셀 값은 0~255범위로, 0~1사이로 정규화 하면 더 안정적이고 빠른 학습 가능
    



### 범주형 데이터
- 수치가 아닌 텍스트 데이터: ['사과','배','포도']처럼 특정 범주를 나타내는 데이터

- 머신러닝 알고리즘은 숫자 데이터를 처리하기 때문에, 범주형 데이터를 숫자로 변환해야 한다.
    1. 레이블 인코딩 방법
    2. 원-핫 인코딩 방법


- 레이블 인코딩
    - 범주형 데이터를 정수로 매핑하여 각 범주를 고유한 숫자로 변환하는 방법
    - 데이터의 범주마다 고유한 번호를 부여 --> 정수 형식으로 출력 

    - 장점
        - 간단하고 직관적이며, 메모리를 적게 사용
    - 단점
        - 범주의 순서가 없는데도 숫자가 순서를 암시할 수 있다.(순서 정보 암시)
    - 사용 시기
        - 범주가 **순서형 데이터**인 경우 적합함

- 원-핫 인코딩
    - 범주형 데이터를 **이진 벡터**로 변환 --> 이진 벡터로 출력
    - 각 범주마다 별도의 열을 만들고, 해당 범주에만 1을 표시하고 나머지는 0으로 표시
    
    - 각 범주를 새로운 열로 추가하고, 데이터가 해당 범주에 속하면 1, 아니면 0으로 표시

    - 장점
        - 숫자가 범주의 순서를 암시하지 않으므로, 레이블 인코딩의 단점을 극복
    - 단점
        - **고차원 문제**: 범주의 수가 많아질수록 많은 열이 생성되어 메모리 소모가 큼(1000개의 범주가 있다면 1000개의 열이 추가됨)
    - 사용 시기
        - 범주가 순서 없는 데이터인 경우 적합함. **비순서형 데이터**

# csv 데이터셋을 통해 원하는 데이터 추출 및 시각화 해보기

```python
#음식점 알아보기
import pandas as pd

data = pd.read_csv('eating_.csv', encoding='cp949', low_memory=False)
data.info()
```

```python
data_1 = data[['번호','영업상태명','사업장명','소재지전체주소','업태구분명','위생업태명']] #업태구분명, 위생업태명: 음식 종류.
data_1 = data_1.dropna(subset=['소재지전체주소'])

data_2 = data_1[data_1['소재지전체주소'].str.contains('서울특별시')]
data_2 = data_2.dropna(subset=['업태구분명'])
```

```python
data_2.head()
```

```python
data_3 = data_2[data_2['업태구분명'] =='한식']
data_3.head()
```

```python
data_4 = data_3[data_3['영업상태명']!='폐업']
```

```python
data_4.head() 
```

```python
data_4.loc[:, '서울시구'] = data_4['소재지전체주소'].str.split().str[1]
data_4.head()
```

```python
data_5 = data_4.groupby('서울시구').size().reset_index(name='사업장수')
data_5
```
